```python
"""
<데이터 수집 형태>
 - 파일 기반 데이터 : 다운로드 방식, 유료or무료(주로 공공기관, 사설기관은 유료가 많음)
 - API 기반 데이터 수집 : 프로그래밍 방식, 유료or무료(주로 공공기관, 사설기관은 유료가 많음)
 - WEB 기반 데이터 수집 : 프로그래밍 방식, 무료(안걸리면..good..걸리면 IP차단됨)  

"""
```

```python
"""
<파일 기반 데이터 수집>
 - 파일 형태 : 주로 excel, csv, json, xml 파일들을 제공해 줍니다.
            : excel은 작은 데이터
            : csv는 많은 데이터
            : json는 key와 value 형태의 딕셔너리 데이터로 특정 포인트의 데이터가 주로 있습니다.(위, 경도 좌표 값 등)
            : xml은 태그기반 데이터로 json 데이터와 유사한 형태의 데이터가 있습니다.(주로 key/value 형태의 데이터)
 
 - 사용되는 라이브러리 : pandas

"""
```

```python
import pandas as pd
```

```python
### bicycle.csv  파일 읽어들이기 
# - 데이터프레임명 : df1 
file_path = "./data/bicycle.csv"
df1 = pd.read_csv(file_path, encoding="euc-kr")
df1
```

```python
### bicycle.xlsx 파일 읽어들이기 
# - 데이터프레임명 : df2
file_path = "./data/bicycle.xlsx"
df2 = pd.read_excel(file_path)
df2
```

```python
### read.json  파일 읽어들이기 
# - 데이터프레임명 : df3
file_path = "./data/read.json"
df3 = pd.read_json(file_path)
df3
```

```python
### 위에 df1, df2, df3의 데이터를 
# - df1_save.csv, df2_save.xlsx, df3_save.json 이름으로 저장하기

df1.to_csv("./data/df1_save.csv")
```

```python
df2.to_excel("./data/df2_save.xlsx")
```

```python
df3.to_json("./data/df3_save.json")
```
