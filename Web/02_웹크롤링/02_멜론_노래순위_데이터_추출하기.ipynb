```python
### 라이브러리 정의
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
### 웹 드라이버 객체 얻어오기
driver = webdriver.Chrome()
```

```python
### 멜론 차트 사이트 접속하기
url = "https://www.melon.com/chart/index.htm"
driver.get(url)
```

```python
### driver 객체가 가지고온 페이지 html 코드 요청하기 : 문자열 타입입니다.
html = driver.page_source
html
```

```python
### 문자열 타입의 태그에 -> 의미 부여 시키기(parser)
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
### 곡정보가 담겨져있는 모든 행(tr 태그)의 태그 추출하기
songs = soup.select("tr")[1:]
print(len(songs))
songs
```

```python
### 0번쨰 곡정보만 추출하여 데이터 추출해보기
song = songs[0]
song
```

```python
### 곡제목 추출하기
# 추출할 패턴 정의
# div 내 class 속성 이름이 ellipsis이면서 rank01인 클래스 이름을 가지는 태그 안에(자식)
# span이라는 태그 바로 밑에(자식) a태그를
pattern = "div.ellipsis.rank01 > span > a"
song.select(pattern)[0].text
```

```python
### 가수 추출하기
# 추출할 패턴 정의
pattern = "div.ellipsis.rank02 > a"
song.select(pattern)[0].text
```

```python
### 곡정보 갯수
len(songs)
```

```python
### 멜론 상위 Top(10개)에 대한 곡정보(곡이름, 가수명) 추출하기
for song in songs[0 : 10]:
    # 곡명
    title = song.select("div.ellipsis.rank01 > span > a")[0].text

    # 가수명
    singer = song.select("div.ellipsis.rank02 > a")[0].text

    msg = f"곡명 : {title} / 가수명 : {singer}"
    print(msg)
```

```python
### 드라이버 종료하기
driver.quit()
```

```python
### BeautifulSoup 정적 데이터 수집
from selenium import webdriver
from bs4 import BeautifulSoup

### 웹 드라이버 객체 얻어오기
driver = webdriver.Chrome()

### 멜론 차트 사이트 접속하기
url = "https://www.melon.com/chart/index.htm"
driver.get(url)

### driver 객체가 가지고온 페이지 html 코드 요청하기 : 문자열 타입입니다.
html = driver.page_source

### 문자열 타입의 태그에 -> 의미 부여 시키기(parser)
soup = BeautifulSoup(html, "html.parser")

### 곡정보가 담겨져있는 모든 행(tr 태그)의 태그 추출하기
songs = soup.select("tr")[1:]

### 멜론 상위 Top(10개)에 대한 곡정보(곡이름, 가수명) 추출하기
for song in songs[0 : 10]:
    # 곡명
    title = song.select("div.ellipsis.rank01 > span > a")[0].text

    # 가수명
    singer = song.select("div.ellipsis.rank02 > a")[0].text

    msg = f"곡명 : {title} / 가수명 : {singer}"
    print(msg)

### 드라이버 종료하기
driver.quit()
```
