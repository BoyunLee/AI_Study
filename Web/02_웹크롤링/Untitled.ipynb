```python
### 라이브러리 정의
# 웹 페이지 컨트롤
from selenium import webdriver

# 웹페이지 내에 태그 및 데이터 추출
from selenium.webdriver.common.by import By

# 페이지 전환 이벤트 발생시(클릭 이벤트) 대기시간을 설정하기 위해 사용
import time
```

```python
driver = webdriver.Chrome()
url = "https://www.megabox.co.kr/movie"
driver.get(url)
```

```python
movie_list_path = "div.movie-list-info"

### 전체 목록에 대한 태그정보 추출하기
# find_element() : 1개 태그 정보 추출시 사용
# find_elements() : 2개 이상의 태그 정보 추출시 사용 -> 리스트 타입으로 반환됨
movie_elements = driver.find_elements(By.CSS_SELECTOR, movie_list_path)

print(len(movie_elements))
```

```python
### 수집년월일
from datetime import datetime

def date(driver):
    date = datetime.now().strftime("%Y-%m-%d")
    return date
```

```python
### 영화제목
def title(driver):
    movie_title_path = "div.movie-detail-cont > p.title"
    
    movie_title_element = driver.find_element(By.CSS_SELECTOR, movie_title_path)
    
    try : 
        title = movie_title_element.text
    except Exception as e:
        title = ""
        
    return title
```

```python
### 예매순위
def booking(driver):
    movie_booking_path = "p.cont > em"
    
    movie_booking_element = driver.find_element(By.CSS_SELECTOR, movie_booking_path)
    
    try : 
        booking = movie_booking_element.text
    except Exception as e:
        booking = ""
    return booking
```

```python
### 예매율
def percent(driver):
    movie_percent_path = "#rkTag"
    
    movie_percent_element = driver.find_element(By.CSS_SELECTOR, movie_percent_path)
    
    try : 
        percent = movie_percent_element.text
    except Exception as e:
        percent = ""
    return percent
```

```python
### 누적관객수
def saw(driver):
    movie_saw_path = "div.audience > p > em"
    
    movie_saw_element = driver.find_element(By.CSS_SELECTOR, movie_saw_path)
    
    try : 
        saw = movie_saw_element.text
    except Exception as e:
        saw = ""
    return saw
```

```python
### 총실관람평점
def stare(driver):
    movie_stare_path = "#mainMegaScore > p > em"
    
    movie_stare_element = driver.find_element(By.CSS_SELECTOR, movie_stare_path)
    
    try : 
        stare = movie_stare_element.text
        if stare == "":
            stare = "0.0"
        
    except Exception as e:
        stare = ""
    return stare
```

```python
date_list =[]
movie_title_list = []
movie_booking_list = []
movie_percent_list = []
movie_saw_list = []
movie_stare_list = []

for i in range(0, 10, 1):
    movie_elements[i].click()
    movie_handle = driver.window_handles[-1]
    driver.switch_to.window(movie_handle)
    time.sleep(1)

    try:
        now_date = date(driver)
        movie_title = title(driver)
        movie_booking = booking(driver)
        movie_percent = percent(driver)
        movie_saw = saw(driver)
        movie_stare = stare(driver)

        date_list.append(now_date)
        movie_title_list.append(movie_title)
        movie_booking_list.append(movie_booking)
        movie_percent_list.append(movie_percent) 
        movie_saw_list.append(movie_saw) 
        movie_stare_list.append(movie_stare) 

    except Exception as e:
        # 요소가 변경되었으므로 다시 찾음
        now_date = date(driver)
        movie_title = title(driver)
        movie_booking = booking(driver)
        movie_percent = percent(driver)
        movie_saw = saw(driver)
        movie_stare = stare(driver)

        date_list.append(now_date)
        movie_title_list.append(movie_title)
        movie_booking_list.append(movie_booking)
        movie_percent_list.append(movie_percent) 
        movie_saw_list.append(movie_saw) 
        movie_stare_list.append(movie_stare) 

    driver.execute_script("window.history.go(-1)")
```
