```python
"""
- df 이름으로 데이터프레임에 모두 담기 
- 컬럼명 : name, price, inventory, stor, url
"""
```

```python
tags_name = soup.select('span.name')
name = tags_name[0].text, tags_name[1].text
name
```

```python
tags_price = soup.select('span.price')
price = tags_price[0].text, tags_price[1].text
price
```

```python
tags_inventory = soup.select('span.inventory')
inventory = tags_inventory[0].text, tags_inventory[1].text
inventory
```

```python
tags_store = soup.select('span.store')
store = tags_store[0].text, tags_store[1].text
store
```

```python
tags_url = soup.select('a')
url = tags_url[0]["href"], tags_url[1]["href"]
url
```

```python
import pandas as pd

# 데이터
data = {
    'name': name,
    'price': price,
    'inventory': inventory,
    'store': store,
    'url' : url
}

# 데이터프레임 생성
df_me = pd.DataFrame(data)

# 결과 확인
df_me
```

```python
df = pd.DataFrame()

for tag in soup.select("p"):
    # print(tag.select(".name")[0].text)
    # print(tag.select(".price")[0].text)
    # print(tag.select(".inventory")[0].text)
    # print(tag.select(".store")[0].text)
    # print(tag.select("a")[0]["href"])
    
    temp_df = pd.DataFrame(columns=["name", "price", "inventory", "store", "url"])
    temp_df["name"] = [tag.select(".name")[0].text]
    temp_df["price"] = [tag.select(".price")[0].text]
    temp_df["inventory"] = [tag.select(".inventory")[0].text]
    temp_df["store"] = [tag.select(".store")[0].text]
    temp_df["url"] = [tag.select("a")[0]["href"]]

    df = pd.concat([df, temp_df], ignore_index=True)
df
```
