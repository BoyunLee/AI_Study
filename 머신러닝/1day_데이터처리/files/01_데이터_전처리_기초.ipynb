### 여러개 파일 통합하기

```python
"""
<통합순서>
1. 데이터 읽어들이기
2. 기준년월 컬럼 추가 : 2019-01 ~ 2021-12
3. 국적명 추가
4. 기존 원본에 행단위 통합
5. 1 ~ 4 반복 (0 ~ 36)
6. 최종 완료된 데이터 파일로 저장

위 순서를 모든 파일에 반복하여 수행
최종 통합본에는 모든 파일의 데이터 전체가 저장되어 있는
데이터 프레임 변수에 저장하여 파일로 저장
"""
```

```python
# zfill 0으로 채우기
yyyy = 2019
mm = 1
str(yyyy) + "-" + str(mm).zfill(2)
```

```python
### 기준년도를 생성하는 반복문 만들기
# sample_1(0)~
# 2019-01 ~ 2021-12
for year in range(2019, 2022):
    for month in range(1, 13):
        print(f"{year}-{str(month).zfill(2)}")
```

```python
### 파일명 반복하면서 생성하기
# sample_1(0) ~ sample_1(36)
for i in range(0, 37):
    file = f"sample_1({i}).xlsx"
    print(file)
```

```python
### 딕셔너리로 년-월 데이터 미리 만들어 놓기
# key는 0~36까지의 번호로 입력
# value는 년-월 값 입력
# 결과 형태 : {0 : '2019-01', 1 : '2019-02', ..., 36 : '2021-12'}

# 딕셔너리 변수
ym_dict = {}

i = 0
for year in range(2019, 2022):
    for month in range(1, 13):
        ym = f"{year}-{str(month).zfill(2)}"
        ym_dict[i] = ym
        i += 1

ym_dict
```

```python
### 전체 통합할 데이터프레임변수 정의
sample_1_all = pd.DataFrame()

for i in range(0, 36):
    file = f"./files2/sample_1({i}).xlsx"
    #print(file, ym_dict[i])
    
    ### 데이터 읽어들이기
    sample_ = pd.read_excel(file_path, header=1, skipfooter=2, usecols="A:C")

    ### 기준년월 컬럼 추가하기
    sample_["기준년월"] = ym_dict[i]

    ### 국적명 추가하기
    sample_code = pd.merge(left = sample_, right = code_master, how = "left", left_on = "국적코드", right_on = "국적코드")
    # print(sample_code)

    sample_1_all = pd.concat([sample_1_all, sample_code], ignore_index=True)

sample_1_all
```
