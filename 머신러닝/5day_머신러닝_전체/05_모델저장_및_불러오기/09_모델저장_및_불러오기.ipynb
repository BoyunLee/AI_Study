```python
"""
- 와인데이터 읽어들이기 : 변수명 wine
- 독립변수 : data와 종속변수 : target으로 분리하기
- 훈련 및 테스트 = 8 : 2로 분류하기(변수는 train_input, train_target, test_input, test_target으로)
- 정규화 하기 : ss 클래스명, 훈련데이터 정규화변수는 train_std_scaler, 테스트데이터는 test_std_scaler로 정규화
- 랜덤포레스트 모델로 분류하기 : 모델 변수명은 rf 사용(cpu core 갯수는 전체사용, 랜덤규칙 42 사용)
- 훈련시키기
"""
```

```python
import pandas as pd

file_path = "./data/08_wine.csv"
wine = pd.read_csv(file_path)
wine
```

```python
import numpy as np
data = wine.iloc[:, :-1].to_numpy()
print(data.shape)

target = wine["class"].to_numpy()
print(target.shape)
```

```python
### 훈련 : 테스트 = 8 : 2로 분류하기
# 변수명 : train_input, train_target, test_input, test_target 사용
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)

print(f"{train_input.shape}, {test_input.shape}, {train_target.shape}, {test_target.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(data)

train_std_scaler = ss.transform(train_input)
test_std_scaler = ss.transform(test_input)

print(f"train_std_scaler = {train_std_scaler}\n test_std_scaler = {test_std_scaler}")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_jobs=-1, random_state=42)
rf
```

```python
rf.fit(train_input, train_target)
```

```python
"""
훈련모델 저장 및 불러오는 방법 2가지
- joblib
- pickle
"""
```

```python
### joblib 방식
```

```python
### joblib 라이브러리
import joblib
```

```python
### 훈련 모델 저장하기
# 저장할 위치 및 파일명 정의 : 확장자는 임의로 넣어도 됩니다.
#  (보통 방식에서는 확장자 를 주로 사용합니다.)
save_path = "./model/rf_model_joblib.md"

# 모델 저장시키기
joblib.dump(rf, save_path)
```

```python
### 저장된 훈련모델 파일 불러들이기
rf_model = joblib.load(save_path)
rf_model
```

```python
### 예측하기
rf_model.predict(test_input)
```

```python
### pickle 방식
```

```python
### pickle 라이브러리
import pickle
```

```python
### 모델 저장하기
save_path = "./model/rf_model_pickle.md"

### wb : w는 파일쓰기 옵션, b는 바이너리 형식 -> 바이너리 형식으로 파일에 쓰기
# pickle방식은 파일을 열어서 쓰는 방식으로 저장됩니다.
with open(save_path, "wb") as fw :
    pickle.dump(rf, fw)
```

```python
### 저장된 모델파일 읽어들이기
with open(save_path, "rb") as f :
    rf_model2 = pickle.load(f)

rf_model2
```

```python
### 예측하기
rf_model2.predict(test_input)
```
