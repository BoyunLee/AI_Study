```python
### 상관관계 표 나타내기
# 사용되는 함수 : corr()
correlation_matrix = df_org.corr()
correlation_matrix
### 표에 보이는 값 -> "상관관계 계수"라고 칭합니다.
# <상관관계 확인하는 방법>
# 1. 왼쪽 상단에서 ~ 오른쪽 하단으로 중앙 대각선을 마음속으로 그립니다.
# 2. 중앙 대각선을 중심으로 오른편 또는 왼편 한쪽한 보시면 됩니다.(서로 대칭)

### 상관관계 계수
# 특성들 간에 선형적인 특성이 있는지에 대한 관계성을 나타낸 지표값
# 지표값을 줄여서 "상관계수(상관관계 계수)" 라고 칭합니다.

### <상관계수 해석방법>
# 상관계수 값은 "절대값"으로 해석합니다.
# 음수값이 있는 상관계수를 "음의 상관관계"라고 칭합니다.
# 양수값이 있는 상관계수를 "양의 상관관계"라고 칭합니다.
# 상관계수 값의 범위 : 0~1(양의 상관관계), 또는 -1~0(음의 상관관계)

# 양의 상관관계 : 양수값(0~1)으로 우상향의 선형관계
# 음의 상관관계 : 음수값(-1~0)으로 우하향의 선형관계

# 해석시에는 절대값으로 해석
# 상관계수의 절대값이 클수록 선형 형태를 잘 나타냅니다.(관계성이 크다고 표현)
# 선형 형태를 "선형성"이라고 칭하기도 합니다.(선형성이 좋다, 나쁘다 / 선형성이 크다, 작다 등으로 표현)
```

```python
X = df_org.iloc[:, :-1]
X
```

```python
"""
독립변수와 종속변수 데이터를 이용하여 -> "상관관계 검정하기"
- 독립변수들의 특성과 종속변수간에 유의미성이 있는지 확인하는 절차로 통계적 기법을 사용합니다.
- 독립변수와 종속변수의 관계가 유의미(p-value)한지 검정하는 단계
    -> p-value값 : 유의미계수라고 칭합니다.
- 유의미한 경우 : p-value < 0.05인 경우를 의미함(유의미성을 가집니다.)
- 유의미하지 않은 경우 : p-value >= 0.05인 경우를 의미함
- 유의미한 기준값 0.05는 정해진 값은 아닙니다.(통계학에서 보편적으로 0.05를 기준으로 하기 때문에 분석에서도 보편적으로 사용)

<검정을 하는 이유>
- 상관계수의 값이 높더라도(선형적이더라도),
  p-value < 0.05의 조건을 만족하지 않는 경우, 즉 유의미하지 않는 경우에는
  해당 특성과의 선형적 특성으로 사용하기에는 어려움이 있는 데이터로 판단합니다.
  -> 즉, 훈련에 적합하지 읺은 데이터(회귀모델에 적용하기 어려운 데이터로 판단합니다.)

- 내가 사용하는 데이터로 연구(분석)을 수행할 수 있을지 없을지를 판단하는 기준값이 됩니다.
- p-value < 0.05 : 유의미함(연구(분석) 가능)
  p-value >= 0.05 : 유의미하지 않음(연구 불가)

- 통계학적으로 사용되는 용어
* 연구를 할 수 있는 경우 : 귀무가설 기각, 대립가설 채택
* 연구를 할 수 없는 경우 : 귀무가설 채택
-> 귀무가설/대립가설은 사전에 정답을 높고 이에 반대되는 논쟁을 할때 사용합니다.
-> 기존의 연구결과는 귀무가설, 이에 반대되는 가설은 대립가설이라고 합니다.
   (보통 연구는 새로운 연구, 기존 연구보다 향상된 연구 등을 수행합니다.)
-> p-value < 0.05 인 경우는 -> 귀무가설을 기각하고 대립가설을 채택하여 연구(분석)이 가능해짐
-> p-value >= 0.05 인 경우는 -> 귀무가설을 채택(기존 연구에서 벗어날 수 없다는 의)
"""
```

```python
"""
<상관관계 검정 방법>
* 스피어만(Spearman) 상관관계 검정 사용목적
- 특성 간의 "비선형적"인 관계도 포함하여 검정에 유용하게 사용되는 검정
- 특성 간의 선형관계의 방향이 변화하지 않고, 일관되게 증가하거나 감소 또는 수평적 관계(선형이 아닌)의 데이터를 검정할 때 주로 사용
- 보통 순위 형태의 데이터 특성 관계가 보일 때 사용(연속성)
  -> 예시 : 시험 성적과 학업 성취도 간의 관계(시험 성적이 높을 수록, 학업 성취도도 높을까?)
   
* 피어슨(Pearson) 상관관계 검정 사용목적
- 특성 간의 "선형적"인 관계를 파악하는데 주로 사용되는 검정
- 일반적으로 주로 사용되는 검정 방법
- 보통 관계를 수치적으로 파악하고자 할 때 사용됨(관계성)
  -> 예시 : 온도와 아이스크림 판매량 간의 관계(날씨가 더우면, 아이스크림 판매량이 증가할까?)
"""
```

```python
### 스피어만(Spearman) 상관관계 검정
# 사용 패키지 : scipy.stats
# 사용 라이브러리 함수 : spearmanr(끝에 r이 하나 들어감)

from scipy.stats import spearmanr
```

```python
### 독립변수의 컬럼명 리스트로 추출하기(반복하기 위해서)
X.columns
```

```python
### 각 특성간의 상관관계 검정하기
# 각각의 특성과 종속변수와의 관계 확인
# 함수의 결과는 2개 반환됩니다.
# -> 첫번째 값(statistic) : 상관계수 값
# -> 두번째 값(p-value) : pvalue 유의미 계수 값

for column in X.columns :
    corr_v, p_value = spearmanr(X[column], y)
    print(f"{column} vs 주택가격 : corr={corr_v:.4f} / p_value={p_value:.4f}")

### 검정 결과의 값에서 p-value의 값이 0.00으로 p-value < 0.05로 면적과 주택가격간에는 유의미함
# 훈련 데이터로 사용가
```

```python
### 피어슨 상관관계 검정하기
from scipy.stats import pearsonr

for column in X.columns :
    corr_v, p_value = pearsonr(X[column], y)
    print(f"{column} vs 주택가격 : corr={corr_v:.4f} / p_value={p_value:.4f}")
```

```python
"""
해석
- 주택가격 예측을 위해 사용하고자 하는 독립변수의 특성들은 모두 p-value < 0.05의 유의미한 결과를 나타냄
- 따라서, 모든 독립변수 특성들을 이용하여 훈련에 사용할 수 있음(연구 진행--->>>)
"""
```

```python
### 독립변수와 종속변수 데이터를 훈련 : 검증 : 테스트 데이터로 분리하기
# 훈련에 사용할 데이터 분리 작업
# 훈련 데이터 : 모델 훈련 및 정확도 확인 시 사용
# 검증 데이터 : 검증 정확도(기존 테스트 정확도) 확인 시 사용
# 테스트 데이터 : 예측 시 사용
# 분류 비율 = 6 : 2 : 2 를 주로 사용하고 있습니다.
# 분류 순서 -> 6(훈련데이터) : 4(검증+테스트 데이터) 분류 -> 4(검증+테스트 데이터)를  5(검증) : 5(테스트)로 분류
# 사용할 변수명 : X_train, X_val, X_test, y_train, y_val, y_test
# 분류 시 사용할 함수 : train_test_split()
```

```python
### 분류 라이브러리
from sklearn.model_selection import train_test_split
```

```python
### 사용할 모델 라이브러리 정의하기
# 선형회귀모델 : 선형방정식을 이용한 방식
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso

# 앙상블 모델 : 트리 분류 구조를 이용한 방식
#            : 여러개의 트리를 사용하는 모델
# 랜덤포레스트 : 여러개의 트리로 훈련 후 가장 선능이 좋은 트리를 사용하는 방식
from sklearn.ensemble import RandomForestRegressor

# 여러개의 트리로 훈련을 하지만, 트리와 트리는 훈련의 선능을 전달 하면서 성능을 계속 높여나가는 구조를 가짐
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import HistGradientBoostingRegressor

### 설치해야 사용가능한 라이브러리 : pip install xgboost
import xgboost as xgb

### 평가 라이브러리 : 회귀분석인 경우에 사용하는 평가방법(분류는 다른 평가방법 사용)
# 모델을 선택할 때 가장 비중을 많이 두는 평가는 R2(결정계수, 설명력, 예측값) 값입니다.
# 평균절대오차(MAE) : 실제값과 예측값 간의 차이를 절대값을 취하여 평균한 값
from sklearn.metrics import mean_absolute_error
# 평균제곱오차(MSE) : 실제값과 예측값 간의 차이를 제곱하여 평균한 값
from sklearn.metrics import mean_squared_error
# 결정계수(R) -> 설명력 = 예측력(훈련결과의 데이터를 잘 설명하고 있는지 여부 확인) : 설명계수
# 모델의 설명력이 좋은지 나쁜지를 나타내는 지표(예측을 잘 할 수 있는지 없는지를 결정)
# 값의 범위는 0~1 : 1에 가까울수록 설명력이 좋다고 표현(분석 보고서에 항상 들어가는 내용)
from sklearn.metrics import r2_score

### 평가시에는 MSE와 R2 평가 결과가 주로 사용됩니다.
# 우선순위는 R2 > MSE > MAE 입니다.

### 모델 선정 시 해석 방법 : 결정계수 값을 이용하여 해석합니다.
# 해당 모델의 결정 계수는 0.XX로 데이터를 잘 설명하고 있음, 즉 설명력이 좋은 모델을 예측 모델로 선정하였음
```

```python
### 훈련 시키기, 정확도 확인, 예측, 평가를 하나의 함수로 정의해 놓기(기능 정의)
# 함수 이름 : train_and_evalute_model

def train_and_evalute_model(model, train_input, train_target, val_input, val_target):
    ### 모델 훈련 시키기
    model.fit(train_input, train_target)

    ### 훈련 및 검증 독립변수로 정확도 확인하기
    train_score = model.score(train_input, train_target)
    val_score = model.score(val_input, val_target)

    ### 훈련 및 검증 데이터를 이용해서 예측 및 평가하기
    # 훈련 예측 평가
    train_pred = model.predict(train_input)
    train_mae = mean_absolute_error(train_target, train_pred)
    train_mse = mean_squared_error(train_target, train_pred)
    train_r2 = r2_score(train_target, train_pred)

    # 검증 예측 및 평가
    val_pred = model.predict(val_input)
    val_mae = mean_absolute_error(val_target, val_pred)
    val_mse = mean_squared_error(val_target, val_pred)
    val_r2 = r2_score(val_target, val_pred)
    
    # print("--------------------------[종료]----------------------------")
    return model, train_score, train_mae, train_mse, train_r2, val_score, val_mae, val_mse, val_r2
```

```python
### 훈련에 사용할 모델 생성하기
liner_reg = LinearRegression()
ridge_reg = Ridge()
lasso_reg = Lasso()
rf_reg = RandomForestRegressor()
gb_reg = GradientBoostingRegressor()
hgb_reg = HistGradientBoostingRegressor()
xgb_reg = xgb.XGBRegressor()

### 리스트에 담아서 반복 수행 할 수 있도록 처리
models = [liner_reg, ridge_reg, lasso_reg, rf_reg, gb_reg, hgb_reg, xgb_reg]
models
```

```python
### 함수 반복 호출하기

results = {}

for m in models :
    ### 모델은 클래스로 되었기 때문에 -> 클래스 이름을 추출 할 수 있습니다.
    model_name = m.__class__.__name__
    print(f"-----------------------------[{model_name}]-------------------------------")

    ### 함수 호출하기
    # print(train_and_evalute_model(m, X_train_scaled, y_train, X_val_scaled, y_val))
    results[model_name] = train_and_evalute_model(m, X_train_scaled, y_train, X_val_scaled, y_val)
    print()

results
```

```python
### 하이퍼파라메터 튜닝 진행하기
# 위에 7개 모델에 대해서 조금 더 성능을 향상시켜서 비교해 볼 수 있는 방법 찾기
# 즉, 하이퍼파라메터 튜닝 진행
# 하이퍼파라메터 튜닝 클래스 사용
# 패키지 : skleatn.model_selection
# 튜닝클래스 : GridSearchCV()

# GridSearchCV()
# 각 모델의 하이퍼파라메터 속성들의 최적의 값들을 스스로 찾아서 훈련까지 수행합니다.
# (하이퍼파라메터 속성 및 속성의 값의 범위를 지정하면, 범위 내에서 가장 적절한 값을 찾아내는 방식을 사용합니다.)
# 즉, 최적의 모델(성능 좋은 모델)을 만들어 냅니다.
# 최적의 모델로 예측하고 평가만 수행하면 됩니다.
# 내부적으로 Cross Validation(CV) 즉, 데이터를 자체적으로 독립변수와 종속변수로 여러 그룹으로 만들어서 
# CV 작업을 -> fold 작업이라고 합니다.
"""
    5 fold
    훈련변수(그룹1 : 그룹2 : 그룹3 : 그룹4) / 검증변수(그룹5)
    훈련련수(그룹1 : 그룹2 : 그룹3 : 그룹5) / 검증변수(그룹4)
    훈련련수(그룹1 : 그룹2 : 그룹4 : 그룹5) / 검증변수(그룹3)
    훈련련수(그룹1 : 그룹3 : 그룹4 : 그룹5) / 검증변수(그룹2)
    훈련련수(그룹2 : 그룹3 : 그룹4 : 그룹5) / 검증변수(그룹1)
"""
# 여러번 교차검증을 스스로 진행하면서 정확도를 높이는 훈련을 수행합니다.
# 시스템 자원을 많이 사용하기 때문에, CPU Core(코어)를 가급적 모두 사용하는 것이 좋습니다.(속도 측면)
```

```python
### CV 갯수 지정하기 : 데이터를 몇개의 그룹으로 분류해서 훈련할지 지정(교차 검증 갯수)
#                     (보통 3 또는 5를 주로 사용합니다.)
cv = 5

### CPU core 갯수 지정하기
# 보통 -1을 사용합니다.(-1은 모든 core를 사용한다는 의미입니다.)
n_jobs = -1

### 평가 방법 지정 : 스스로 평가까지 진행시킵니다.(가장 좋은 성능을 만들기 위해서)
# neg_mean_squared_error : 평균제곱오차(MSE)를 의미합니다.
# r2 :결정계수 r2_score()를 의미합니다.
scoring = ["neg_mean_squared_error", "r2"]

### 모델 선정기준 지정 : 최종 모델 선정시 결정 기준을 평가방법 중에 지정합니다.
refit = "r2"
```

```python
### 하이퍼파라메터 튜닝을 위한 함수 생성하기
# 함수 이름 : get_HyperParameter_Tuning

def get_HyperParameter_Tuning(model, gridParam, scoring, refit, X_train, y_train, X_val, y_val, cv=5, n_jops=-1):

    ### 튜닝객체 생성하기
    grid_search_model = GridSearchCV(model, gridParams, scoring=scoring, refit=refit, cv=cv, n_jobs=n_jobs)

    ### 튜닝객체 훈련시키기(모델 훈련시키기)
    grid_search_model.fit(X_train_scaled, y_train)

    ### 최적의 모델을 변수로 받아오기
    best_model = grid_search_model.best_estimator_

    ### 훈련데이터로 예측 및 평가결과 확인하기
    # 훈련 예측하기
    train_pred = best_model.predict(X_train)
    # 훈련 MAE
    train_mae = mean_absolute_error(y_train, train_pred)
    # 훈련 MSE
    train_mse = mean_squared_error(y_train, train_pred)
    # 훈련 r2
    train_r2 = r2_score(y_train, train_pred)

    ### 검증데이터로 예측 및 평가결과 확인하기
    # 검증 예측하기
    val_pred = best_model.predict(X_val)
    # 검증 MAE
    val_mae = mean_absolute_error(y_val, val_pred)
    # 검증 MSE
    val_mse = mean_squared_error(y_val, val_pred)
    # 검증 r2
    val_r2 = r2_score(y_val, val_pred)

    return best_model, train_mae, train_mse, train_r2, val_mae, val_mse, val_r2
```

```python
### 훈련에 사용할 모델 생성하기
liner_reg = LinearRegression()
ridge_reg = Ridge(alpha=1)
lasso_reg = Lasso(alpha=1)
rf_reg = RandomForestRegressor()
gb_reg = GradientBoostingRegressor()
hgb_reg = HistGradientBoostingRegressor()
xgb_reg = xgb.XGBRegressor()

### 리스트에 담아서 반복 수행 할 수 있도록 처리
models = [liner_reg, ridge_reg, lasso_reg, rf_reg, gb_reg, hgb_reg, xgb_reg]
models
```

```python
### 튜닝 함수 호출하기
# 각 모델 리턴 결과 저장할 딕셔너리 변수 정의
results = {}

# 각 모델의 실행결과를 데이터프레임에 담을 변수 정의
rs_df = pd.DataFrame()

for m in models :
    ### 모델 이름 추출
    modelName = m.__class__.__name__
    print(f"---------------------[{modelName}]----------------------")

    ### 하이퍼파라메터 속성이름과 값 범위 지정하기
    # 모델마다 속성이름이 다르기에 조건식으로 정의해야함

    ### 하이퍼파라메터를 담을 딕셔너리 변수 정의
    gridParams = {}
    
    ### LinearRegression, Ridge, Lasso 모델인 경우
    # 선형회귀모델들은 특별한 튜닝변수가 없습니다. 형식상 사용해보겠습니다.
    if modelName in ["LinearRegression", "Ridge", "Lasso"] :
        # 선형방정식에서 절편 사용여부(모델 기본값 True)
        gridParams["fit_intercept"] = [True, False]
        # 예측값을 양수로 사용할지 여부(모델 기본값 False)
        gridParams["positive"] = [True, False]
    
    ### GradientBoostingRegressor인 경우
    elif modelName == "GradientBoostingRegressor" :
        # 트리 갯수 지정(보통 50~100 사이값 사용)
        gridParams["n_estimators"] = [50, 100]
        # 트리가 뻗어나가는 최대 깊이 지정(None은 제안두지 않음, 보통 3~10 사이값 사용)
        gridParams["max_depth"] = [None, 3, 10]
        # 학습율(보폭) : 0.01~0.2 사이값 사용
        gridParams["learning_rate"] = [0.01, 0.1, 0.2]
        
    ### HGB인 경우
    elif modelName == "HistGradientBoostingRegressor" :
        # 트리 갯수 지정(보통 50~100 사이값 사용)
        gridParams["max_iter"] = [50, 100]
        # 트리가 뻗어나가는 최대 깊이 지정(None은 제안두지 않음, 보통 3~10 사이값 사용)
        gridParams["max_depth"] = [None, 3, 10]
        # 학습율(보폭) : 0.01~0.2 사이값 사용
        gridParams["learning_rate"] = [0.01, 0.1, 0.2]
    
    ### RF 및 XGB인 경우
    else:
        # 트리 갯수 지정(보통 50~100 사이값 사용)
        gridParams["n_estimators"] = [50, 100]
        # 트리가 뻗어나가는 최대 깊이 지정(None은 제안두지 않음, 보통 3~10 사이값 사용)
        gridParams["max_depth"] = [None, 3, 10]

    ### 평가방법
    scoring = ["neg_mean_squared_error", "r2"]

    ### 모델선정방법(기준값) 지정
    refit = "r2"

    ### 함수 호출하기 -----------------------------
    best_model, train_mae, train_mse, train_r2, val_mae, val_mse, val_r2 = get_HyperParameter_Tuning(m, gridParams, scoring, refit, 
                                                                                                     X_train_scaled, y_train, X_val_scaled, y_val)
    # print(best_model, train_mae, train_mse, train_r2, val_mae, val_mse, val_r2)

    ### 딕셔너리에 모델 별로 결과값 담기
    results[modelName] = [best_model, train_mae, train_mse, train_r2, val_mae, val_mse, val_r2]

    ### 데이터프레임에 모델 별로 결과값 담기
    rs_temp = pd.DataFrame([[modelName, train_mae, train_mse, train_r2, val_mae, val_mse, train_r2 - val_r2]],
                              columns=["modelName", "train_mae", "train_mse", "train_r2", "val_mae", "val_mse", "train_r2 - val_r2"])

    ### 원본 데이터프레임에 rs_temp 데이터를 행단위로 추가하기
    rs_df = pd.concat([rs_df, rs_temp], ignore_index=True)
```
