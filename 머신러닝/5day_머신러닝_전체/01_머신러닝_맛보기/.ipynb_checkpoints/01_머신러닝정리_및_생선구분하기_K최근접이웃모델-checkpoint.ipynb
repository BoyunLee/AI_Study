```python
"""
<생선 구분하기>
- 특정(컬럼) : 생선이름, 생선길이, 생선무게
- 길이와 무게로 생성이름 맞추기

- 독립변수(x) = 길이와 무게
- 종속변수(y) = 이름(도미와 방어 구분하기)

- 종속변수가 범주형 데이터, 즉 분류하는 모델 적용

<훈련 모델 처리 순서> - 머신러닝, 딥러닝
1. 데이터 수집
2. 데이터 전처리 : 표준화(정규화) -> 특성(컬럼)들간의 데이터 단위(scale)가 다른 경우 일정한 크기로 맞춰주는 처리
3. 데이터 분류 : 훈련(7:3 or 7.5:2.5 or 8:2)/테스트 또는 훈련/검증/테스트(6:2:2)로 데이터 분류
4. 모델 생성 : 모델은 클래스 객체로 라이브러리화 되어 있음(모델 생성이라는 의미는 클래스 생성을 의미함)
5. 모델 훈련(fit)
6. 모델 평가
7. 모델 튜닝(하이퍼파라메터 튜닝) : 하이퍼파라메터를 속성들이라고 생각하면 됨
8. 5 ~ 7번 반복
9. 모델 선정 : 분류모델 전체를 훈련 후 평가 결과가 가장 좋은 모델을 선정합니다.
10. 선정모델로 최종 테스트(predict)
"""
```

```python
### 데이터 불러들이기
# 도미의 길이와 무게 데이터
# 도미 길이(cm)
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 
                30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 
                33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 
                39.5, 41.0, 41.0]
# 도미 무게(g)
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 
                390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 
                600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 
                685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 
                850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

```

```python
### 도미의 길이와 무게의 데이터 길이 확인하기
len(bream_length), len(bream_weight)
```

```python
### 데이터 분포 확인하기
### 시각화 라이브러리
# 산점도 그래프 그리기 : scatter()
import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight)
# x축 이름
plt.xlabel("length")
# y축 이름
plt.ylabel('weight')
plt.show()

### 해석
# 아래 그래프는 길이가 증가할수록 무게도 증가는 양상을 보이고 있다.
# 산점도의 분포 방향은 양의 방향을 띄고 있다. 이를 양의 상관관계라고 한다.
# 양의 상관관계를 띄는 우상향 분포를 나타내고 있다.
# 우상향 형태의 선형 형태를 띄고 있음
```

```python
### 데이터 불러들이기
# 빙어의 길이와 무게 데이터
# 빙어 길이(cm)
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
# 빙어 무게() 
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
### 빙어의 길이와 무게의 데이터 길이 확인하기
len(smelt_length), len(smelt_weight)
```

```python
### 데이터 분포 확인하기
### 시각화 라이브러리
# 산점도 그래프 그리기 : scatter()
import matplotlib.pyplot as plt
plt.scatter(smelt_length, smelt_weight)
# x축 이름
plt.xlabel("length")
# y축 이름
plt.ylabel('weight')
plt.show()

### 해석
# 아래 그래프는 길이가 증가할수록 무게도 증가는 양상을 보이고 있다.
# 산점도의 분포 방향은 양의 방향을 띄고 있다. 이를 양의 상관관계라고 한다.
# 양의 상관관계를 띄는 우상향 분포를 나타내고 있다.
# 우상향 형태의 선형 형태를 띄고 있음
```

```python
### 도미와 빙어 데이터를 하나의 산점도 그래프로 그리기
# 도미는 빨강색으로 표시, 빙어는 파랑색으로 표시
# 범례도 추가

import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight, c='red', label='bream')
plt.scatter(smelt_length, smelt_weight, c='blue', label='smelt') 
# x축 이름
plt.xlabel('length')
# y축 이름
plt.ylabel('weight')
# 범례 표시하기
plt.legend()
plt.show()

### 해석
# 도미와 빙어는 모두 우상향 분포를 나타내고 있으며,
# 도미의 경우 경사가 급한반면, 빙어의 경우에는 경사(기울기)가 와만한 분포를 보이고 있음
```

```python
### 생선 분류하기 | 분류모델 적용 -> 도미와 방어 분류하기
```

```python
"""
<KNN 분류모델 사용>
- K 최근접이웃 모델이라고 칭합니다.
- 머신러닝에서 가장 간단한 모델로 처음에 비교용으로 주로 사용되는 모델
- 실제 서비스용으로 사용하기에는 성능이 약한 모델

<KNN 처리방식>
- 분류하고자 하는 값과 가장 가까운 이웃 갯수의 비율(또는 평균)을 이용하여,
  비율이 가장 많은 쪽(다수결의 원칙을 따름)으로 판단하여 분류하는 방식
- 분류하고자 하는 값과 가장 가까운 이웃들을 K라고 합니다.
- K의 값은 분류시 기준이되는 이웃의 갯수를 의미
"""
```

```python
"""
<훈련에 사용할 데이터>
- x(독립변수) : 길이, 무게
- y(종속변수) : 이름

- 독립변수 도미와 빙어 데이터를 하나로 통합해야함
- 종속변수는 생성해야함
"""
```

```python
### 도미와 빙어 데이터 통합하기(1차원)
# 도미와 빙어의 길이 데이터 합치기
length = bream_length + smelt_length

# 도미와 빙어의 무게 데이터 합치기
weight = bream_weight + smelt_weight

len(length), len(weight)
```

```python
print(length,'\n\n',weight)
```

```python
"""
< 머신러닝에서 사용하는 차원>
- 독립변수 : 2차원
           : [[길이, 무게], [길이, 무게], [길이, 무게], [], ...]
- 종속변수 : 1차원
           : [이름, 이름, 이름, 이름, ...]
"""
```

```python
# 2차원 만들기(1)
fish_data = []
for i in range(0, len(length), 1):
    fish_data.append([length[i], weight[i]])
fish_data    

# 2차원 만들기(2)
fish_data = []
for l, w in zip(length, weight):
    fish_data.append([l, w])
fish_data 

# 2차원 만들기(3)
fish_data = [[l, w] for l, w in zip(length, weight)]
fish_data 
```

```python
"""
<종속변수 데이터 생성하기>
- 1차원으로 만들어야 합니다.
- 49개의 이름은 독립변수의 인덱스 위치의 이름과 동일하게 만들어야 합니다.
- 머신러닝에서는 숫자값을 사용합니다.
- 도미는 1로 빙어는 0으로 생성
- 이진분류 : 0과 1의 값을 사용
- 다중분류 : 0~n의 값을 사용
"""
```

```python
len(bream_length), len(smelt_length)
```

```python
### 종속변수 데이터 생성
fish_target = [1] * 35 + [0] * 14
print(len(fish_target), "\n\n", fish_target)
```

```python
"""
모델 생성하기
- 사용하는 모델 : KNeighbors의 분류모델 -> KNeighborsClassifier
"""
```

```python
### 모델 라이브러리 읽어들이기
from sklearn.neighbors import  KNeighborsClassifier
```

```python
### 분류모델(클래스) 생성시키기
kn = KNeighborsClassifier()
kn
```

```python
"""
모델 훈련(학습) 시키기
- 지도학습인 경우 : fit(독립변수, 종속변수)
- 비지도학습인 경우 : fit(독립변수)
"""
```

```python
kn.fit(fish_data, fish_target)
```

```python
"""
정확도 확인하기
- 분류에서 정확도는 : score(독립변수, 종속변수) 함수 사용
"""
```

```python
### 분류 정확도
kn.score(fish_data, fish_target)

### 해석
# 분류에서의 정확도 범위 : 0~1 (0% ~ 100%)사이의 값이 출력됨
# 1에 가까울수록 정확도가 높다고 판단
# 즉, 성능이 좋은 훈련모델이라고 판단
```

```python
### 임의 데이터를 이용해서 분류해보기(테스트)
# 예측함수 : pcredit(임의 독립변수) 사용
kn.predict([[30, 600]])

### 해석
# 출력결과 값이 "종속변수"값 - 예측값(분류값)이라고 칭함
```

```python
### 빙어, 도미, 임의 데이터를 산점도로 그리기
# 도미는 빨강색, 빙어는 파랑색, 임의 데이터는 초록색으로 표시
# 범례도 추가

import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight, c='red', label='bream')
plt.scatter(smelt_length, smelt_weight, c='blue', label='smelt') 
plt.scatter(30, 600, marker='^', c='green', label='pred') 
# x축 이름
plt.xlabel('length')
# y축 이름
plt.ylabel('weight')
# 범례 표시하기
plt.legend()
plt.show()

### 해석 
# 임의의 예측 데이터가 위치한 곳의 이웃의 갯수를 확인하여
# 비율이 높은 쪽의 이름으로 분류하게 됨
# K(이웃의 갯수)는 기본(default)로 5개가 사용됨(지정 안할 경우)
```

```python
### 하이퍼파라메터 튜닝
# 모델 성능을 향상시키기 위한 방법
# 과대적합 또는 과소적합이 일어난 경우 튜닝 진행
# 모델(클래스)의 속성(변수) 중에 사람이 직접 값을 지정해 주어야하는 변수들을 통칭해서 '하이퍼파라메터'라고 칭합니다.

# 과대적합 : 정확도가 1인 경우 
```

```python
### KNN 분류 모델 생성하기
# n_neighbors : 이웃의 갯수(하이퍼파라메터)
kn21 = KNeighborsClassifier(n_neighbors=21)
kn21
```

```python
### KNN 분류모델 훈련시키기
kn21.fit(fish_data, fish_target)
```

```python
### 훈련모델 정확도 확인하기
kn21.score(fish_data, fish_target)

### 해석
# 도미와 빙어의 무게와 길이 데이터를 이용한 분류시
# KNN 분류모델의 이웃의 갯수(K)를 21개로 사용하였을 때
# 정확도 0.98의 매우 우수한 모델로 판단된다.
```

```python
### 임의 데이터 30 길이의 600무게로 분류해보기
kn21.predict([[30, 600]])

### 임의 독립변수를 이용한 예측 결과는 1로 도미로 분류 되었음
```

```python
### 정확도가 1보다 작은 가장 좋은 정확도일 때의 이웃의 갯수 찾기
### 모델 생성
kn = KNeighborsClassifier()

### 훈련시키기
kn.fit(fish_data, fish_target)

### 다음부터는 가장 회적의 이웃을 찾는 로직 만들어주세요
# 내가 적은 답(틀림)
max1 = 0
for i in range(3, 49, 2):
    kn.n_neighbors = i
    kn.fit(fish_data, fish_target)
    score = kn.score(fish_data, fish_target)
    if score > max1:
        max1 = score
        if max1 >= 1:
            print(i - 1)
            break
```

```python
### 다음부터는 가장 회적의 이웃을 찾는 로직 만들어주세요

# 정답
# 반복문의 범위 : 최소 3부터 최대 데이터 갯수만큼
# 단, 최대 데이터의 갯수는 실제 모두 사용하는 경우가 없기에 1/2 정도만 사용해도 됨

# 정확도가 가장 높을 때의 이웃의 갯수를 담을 변수
nCnt = 0

# 정확도가 가장 높을 때의 정확도를 담을 변수
nScore = 0

for n in range(3, len(fish_data), 2):
    kn.n_neighbors = n
    # 이미 모델 객체가 생성되어 있는 경우에는 score()만 가지고도 정확도 확인 가능
    # fit은 생략가능 -> score()함수 내에서도 fit을 자동 수행하고 있음
    kn.fit(fish_data, fish_target)
    score = kn.score(fish_data, fish_target)
    
    # print(n, score)
    # 1보다 작은 정확도인 경우만 처리
    if score < 1:
        # 담아놓은 nScore값이 현재 Score보다 작은 경우에는 현재 Score로 교체하기
        if nScore < score:
            nScore = score
            # 이때의 n(이웃의 갯수)를 nCnt 변수에 담아 놓기
            nCnt = n
            
print(f"nCnt={nCnt}, nScore={nScore}")
```

```python
### 찾아낸 이웃의 갯수 변수를 이용해서 최종 테스트
kn.n_neighbors = nCnt
kn.score(fish_data, fish_target)

kn.predict([[30, 600]])
```
