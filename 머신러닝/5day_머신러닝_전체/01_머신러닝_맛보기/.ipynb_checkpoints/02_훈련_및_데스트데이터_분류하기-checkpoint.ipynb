```python
"""
<모델 훈련을 위한 데이터 구성 방법>
1. 수집한 데이터를 이용해서 훈련:테스트 또는 훈련:검증:테스트 데이터로 분류
2. 분류 기준 : 훈련:테스트 -> 주로 사용되는 비율은 7:3(주로사용) 또는 7.5:2.5
             : 훈련:검증:테스트 -> 주로 사용되는 비율은 6:2:2(주로사용) 또는 5:2.5:2.5
3. 변수 이름 정의(별도로 정의된 이름은 없지만 주로 많이 사용되는 이름)
 - 훈련데이터 : 훈련(fit)시에 사용되는 데이터
              : 독립변수 이름 : train_input, train_x, X_train
              : 종속변수 이름 : train_target, train_y, Y_train
 - 검증데이터 : 정확도(score)시에 사용되는 데이터
              : 독립변수 이름 : val_input, val_x, X_val
              : 종속변수 이름 : val_target, val_y, Y_val
 - 데스트데이터 : 예측(predict)시에 사용되는 데이터
                : 독립변수 이름 : test_input, test_x, X_test
                : 종속변수 이름 : test_target, test_y, Y_test

<데이터 분류 순서>
1. 훈련:테스트 인 경우에는 비율로 분류한 하면 끝
2. 훈련:검증:테스트 인 경우 (6:2:2 인 경우)
    - (훈련+검증)과 테스트 데이터로 8:2 분류
    - (훈련+검증) 데이터를 다시 훈련:검증으로 6:2로 분류
"""
```

```python
### 사용할 데이터 정의하기
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

len(fish_length), len(fish_weight)
```

```python
### 독립변수 만들기
# 사용할 변수명 fish_data
fish_data = [[l, w]for l, w in zip(fish_length, fish_weight)]
len(fish_data)
print(fish_data)
```

```python
### 종속변수 만들기
# 사용할 변수명 fish_target
fish_target = [1]*35 + [0]*14
len(fish_target)
print(fish_target)
```

```python
### 훈련 및 테스트 데이터로 분류하기
```

```python
### 훈련데이터(train) 만들기
# fish_data 및 fish_target의 앞쪽 35개를 독립/테스트로 사용
# 변수명 : train_input, train_target

### 훈련 독립변수
train_input = fish_data[:35]

### 훈련 종속변수
train_target = fish_target[:35]

print(train_input)
print(train_target)
```

```python
### 테스트데이터 만들기
# fish_data 및 fish_target의 뒤쪽 14개를 독립/테스트로 사용
# 변수명 : test_input, test_target

### 훈련 독립변수
test_input = fish_data[35:]

### 훈련 종속변수
test_target = fish_target[35:]

print(test_input)
print(test_target)
```

```python
### 훈련시키기
```

```python
### 모델 생성하기(KNN 분류 모델 사용)
# 모델 변수명 : kn 사용 (이웃의 갯수는 디폴트 사용)
from sklearn.neighbors import  KNeighborsClassifier
kn = KNeighborsClassifier()
kn
```

```python
### 모델 훈련 시키기 (fit()함수 사용)
kn.fit(train_input, train_target)
```

```python
### 모델 정확도 확인하기 (score()함수 사용)
# 훈련 정확도 (훈련 데이터 사용)
train_score = kn.score(train_input, train_target)
train_score
```

```python
# 테스트 정확도 (테스트 데이터 사용)
test_score = kn.score(test_input, test_target)
test_score
```

```python
"""
<테스트 데이터가 0점이 나온 원인>
- 훈련과 테스트 데이터 분류시에 도미 35개를 훈련으로, 빙어 14를 테스트로 분리하여
  모델 훈련시에 훈련데이터(도미 35)로 훈련된 상태에서
  테스트(빙어 14)를 맞추려고 하다보니 훈련이 되어 있지 않아서 발생한 문제
- 즉, 훈련 데이터가 잘못 구성되어 있음
- 훈련과 테스트 데이터 어느 한쪽에 특정 데이터 성격만 편중된 경우에 발생
- 이러한 현상을 '샘플링편향'이라고 함

<샘플링 편향 해소방법>
- 훈련 및 테스트 데이터로 분류하기 전에 데이터를 잘 섞어야합니다.(셔플이라고 함)
"""
```

```python
### 샘플링 편향 해소하기
```

```python
# numpy 라이브러리 사용하기
import numpy as np
```

```python
"""
<리스트[]와 Numpy 배열의  array([])차이점>
- 파이썬 리스트 : 다양한 타입의 값들을 넣을 수 있음(문자, 숫자, 객체 모두 사용가능)
- Numpy 배열 : 한가지 타입의 데티어만 넣을 수 있음(숫자 값 사용가능)
"""
```

```python
### 넘파이 배열 형태를 이용해서 처리하기
# 기존의 독립 및 종속 데이터를 numpy의 배열 타입으로 형변환 시키기
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)
input_arr, target_arr

# 각 배열의 객수 확인하기
len(input_arr), len(target_arr)

# 각 배열의 차원 확인하기
input_arr.shape, target_arr.shape,
```

```python
### 랜덤하게 섞기
# 랜덤 규칙 정의하기 : 다른 사람과 훈련에 대한 정확도를 비교하기 위해서 주로 사용
# seed() 함수 사용 : 사용되는 값은 의미 없는 값임, 규칙을 위한 값(주로 42번 많이 사용)

np.random.seed(42)

### 독립변수와 종속변수의 인덱스 위치를 동일하게 맞춰서 섞어야합니다.
# 랜덤한 인덱스 49개 만들기
# 배열 49개 숫자 만들기
index = np.arange(49)
index

# 랜덤하게 섞기
np.random.shuffle(index)
index
```

```python
### 훈련 및 테스트 데이터 분류하기
# 훈련 및 테스트 독립변수 : input_arr 데이터에서 index 앞쪽 35개, 뒤쪽 14개 값을 인덱스 번호로 사용하여 추출하기
# 훈련 및 테스트 종속변수 : target_arr 데이터에서 index 앞쪽 35개, 뒤쪽 14개 값을 인덱스 번호로 사용하여 추출하기
# 훈련 데이터
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

# 테스트 데이터
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
### 훈련 및 테스트 데이터 차원 확인하기
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
### 훈련 및 테스트 데이터를 이용해서 산점도 그리기
import matplotlib.pyplot as plt

# 훈련 독립변수 데이터를 이용해서 길이와 무게 산점도 표시하기
plt.scatter(train_input[:, 0], train_input[:, 1], c = 'red', label = 'train')

# 테스트 독립변수 데이터를 이용해서 길이와 무게 산점도 표시하기
plt.scatter(test_input[:, 0], test_input[:, 1], c = 'blue', label = 'test')

plt.xlabel("length")
plt.ylabel("weight")
plt.legend()
plt.show()
```

```python
### 이웃의 갯수 튜닝하기
### 정확도가 1보다 작은 가장 좋은 정확도일 때의 이웃의 갯수 찾기

# 반복문의 범위 : 최소 3부터 최대 데이터 갯수만큼
# 단, 최대 데이터의 갯수는 실제 모두 사용하는 경우가 없기에 1/2 정도만 사용해도 됨

# 정확도가 가장 높을 때의 이웃의 갯수를 담을 변수
nCnt = 0

# 정확도가 가장 높을 때의 정확도를 담을 변수
nScore = 0

for n in range(3, len(train_input), 2):
    kn.n_neighbors = n
    # 이미 모델 객체가 생성되어 있는 경우에는 score()만 가지고도 정확도 확인 가능
    # fit은 생략가능 -> score()함수 내에서도 fit을 자동 수행하고 있음
    kn.fit(train_input, train_target)
    score = kn.score(train_input, train_target)
    
    # print(n, score)
    # 1보다 작은 정확도인 경우만 처리
    if score < 1:
        # 담아놓은 nScore값이 현재 Score보다 작은 경우에는 현재 Score로 교체하기
        if nScore < score:
            nScore = score
            # 이때의 n(이웃의 갯수)를 nCnt 변수에 담아 놓기
            nCnt = n
            
print(f"nCnt={nCnt}, nScore={nScore}")
```

```python
### 훈련과 테스트 데이터로 7:3 분류하기
# 첫번째 인자 : 독립변수 원본 데이터
# 두번째 인자 : 종속변수 원본 데이터
# test_size : 훈련:테스트의 비율중에 test의 비율을 작성(1을 기준으로 비율 작성)
# radom_state : 랜덤 규칙 번호 지정
# stratify : 종속변수의 범주 비율을 훈련과 테스트의 비율대비 편향 없이 조정시킵니다.
# 결과값 : 훈련독립, 테스트독립, 훈련종속, 테스트종속의 순서로 데이터가 분류됨

train_input, test_input, train_target, test_target = train_test_split(fish_data_arr, fish_target_arr, 
                                                                     test_size=0.3, random_state=42, stratify=fish_target)
```

```python
"""
<해석>
- 해석은 과대적합과 과소적합이라는 용어를 이용해서 모델 성능을 평가 합니다.

(과대적합)
- 훈련 정확도 > 테스트(or 검증) 정확도 또는 훈련정확도가 1인 경우
    (훈련 정확도가 너무 많이 클경우)

(과소적합)
- 훈련 정확도 < 테스트(or 검증) 정확도 또는 테스트(or 검증) 정확도가 1인 경우
    (훈련 정확도는 테스트 정확도보다 무조건 작으면 안됨)
- 과대 또는 과소 적합이 일어난 경우에는 훈련모델 성능 향상이 필요하다고 판단합니다.
    즉, 모델을 사용할 수 없음

(모델 선정 기준)
- 과소적합이 일어나지 않으면서
- 훈련 정확도가 1이 아니고
- 훈련과 테스트(or 검증) 정확도 차이가 보통(기준은 아님) 0.1 이내인 경우

* 선정된 모델은 "일반화 모델"이라고 칭합니다.
* 가장 바람직한 크기 비교 : 훈련 > 검증 > 테스트 
"""
```

```python
# 정확도가 가장 높을 때의 이웃의 갯수를 담을 변수
nCnt = 0

# 정확도가 가장 높을 때의 정확도를 담을 변수
nScore = 0

for n in range(3, len(train_input), 2):
    kn.n_neighbors = n
    # 이미 모델 객체가 생성되어 있는 경우에는 score()만 가지고도 정확도 확인 가능
    # fit은 생략가능 -> score()함수 내에서도 fit을 자동 수행하고 있음
    kn.fit(train_input, train_target)
    score = kn.score(train_input, train_target)
    
    # print(n, score)
    # 1보다 작은 정확도인 경우만 처리
    if score < 1:
        # 담아놓은 nScore값이 현재 Score보다 작은 경우에는 현재 Score로 교체하기
        if nScore < score:
            nScore = score
            # 이때의 n(이웃의 갯수)를 nCnt 변수에 담아 놓기
            nCnt = n
            
print(f"nCnt={nCnt}, nScore={nScore}")
```

```python
"""
해석
- 예측 결과는 도미로 확인 되었으나, 표시된 산점도를 볼 때 빙어의 이웃만 사용되었음
- 즉, 이웃의 과반수가 빙어 쪽이 많기에 결과값은 빙어가 나와야 할 것 같은데? 왜~왜~?

- 이런 현상이 발생하는 경우는 보통 사용된 특성들의 단위(scale)이 다른 경우 주로 발생 합니다.
    ('훈련데이터의 스케일이 다르다'라고 표현합니다.)

- 해소 방법 : 데이터의 스케일을 '표준화(정규화)'하는 전처리를 수행해야합니다. 
"""
```

```python
"""
<표준화(정규화) 방법>
- 표준점수를 계산하여 사용 가능
- 표준점수 = (각 데이터 - 데이터 전체 평균) / 데이터 전체 표준편차
- 표준점수란 : 각 데이터가 원점(0)에서 표준편차 만큼 얼마나 떨어져 있는지를 나타낸 값

<정규화(표준화)를 수행하는 시점>
1. 데이터 수집
2. 독립데이터와 종속데이터로 분리
3. 독립데이터에 대한 -> 정규화(표준화) 처리 가능(1)
4. 훈련:검즘:테스트 데이터로 분류
5. 훈련, 검증, 테스트 독립데이터에 대한 -> 정규화(표준화) 처리 가능(2)
6. 훈련 모델 생성
7. 훈련 시키기(fit)
8. 정확도 확인(score)
9. 과대/과소 적합 여부 확인
10. 하이퍼파라메터 튜닝
11. 6~10 반복 수행
12. 최종 테스트(predict)
"""
```

```python
# 표준점수 = (각 데이터 - 데이터 전체 평균) / 데이터 전체 표준편차

### 훈련 독립변수에 대한 전체 평균 구하기
mean = np.mean(train_input, axis=0)
print(mean)

### 표준편차 구하기
std = np.std(train_input, axis=0)
print(std)

### 정규화(표준화) 처리하기
train_scaled = (train_input - mean) / std
train_scaled
```

```python
train_input
```

```python
### 사용된 이웃을 포함하여 훈련 독립변수 데이터를 이용하여 산점도 그리기
# 임의 데이터도 함께 표시하기
plt.scatter(train_scaled[:, 0], train_scaled[:, 1], c = 'red', label = 'train')
plt.scatter(25, 150, marker='^', c = 'green', label = 'pred')

# 사용된 이웃 표시하기
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], c = 'blue', label = 'kn')

plt.xlabel("length")
plt.ylabel("weight")
plt.legend()
plt.show()

### 임의 값(25, 150) 값만 다른 위치에 표시됨
# 임의 테스트 데이터도 정규화 작업을 수행 해야함
```

```python
### 예측하고자 하는 임의 데이터 정규화 하기
new = ([25, 150] - mean) / std
new
```

```python
### 사용된 이웃을 포함하여 훈련 독립변수 데이터를 이용하여 산점도 그리기
# 임의 데이터도 함께 표시하기
plt.scatter(train_scaled[:, 0], train_scaled[:, 1], c = 'red', label = 'train')
plt.scatter(new[0], new[1], marker='^', c = 'green', label = 'pred')

# 사용된 이웃 표시하기
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], c = 'blue', label = 'kn')

plt.xlabel("length")
plt.ylabel("weight")
plt.legend()
plt.show()

### 임의의 값(25, 150) 값만 다른 위치에 표시됨
# 임의 테스트 데이터도 정규화 작업을 수행 해야함
```

```python
### 정규화 데이터로 훈련 다시 수행하기
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)
kn.score(train_scaled, train_target)
```

```python
### 테스트 데이터 정확도 확인
# 훈련데이터의 형태를 그대로 따라야함 : 테스트 데이터도 정규화(표준화) 전처리 해야 합니다.
test_scaled = (test_input - mean) / std
kn.score(test_scaled, test_target)
```
