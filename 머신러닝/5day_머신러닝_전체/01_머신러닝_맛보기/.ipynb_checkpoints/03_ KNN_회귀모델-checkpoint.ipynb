```python
"""
<정규화(표준화)를 수행하는 시점>
1. 데이터 수집
2. 독립데이터와 종속데이터로 분리
3. 독립데이터에 대한 -> 정규화(표준화) 처리 가능(1)
4. 훈련:검즘:테스트 데이터로 분류
5. 훈련, 검증, 테스트 독립데이터에 대한 -> 정규화(표준화) 처리 가능(2)
6. 훈련 모델 생성
7. 훈련 시키기(fit)
8. 정확도 확인(score)
9. 과대/과소 적합 여부 확인
10. 하이퍼파라메터 튜닝
11. 6~10 반복 수행
12. 최종 테스트(predict)
"""
```

```python
### 농어 길이를 독립변수로 무게를  종식변수로 정의하여 훈련 시킨 후,
# 임의 테스트 값(길이)를 이용하여 (무게)를 예측하는 모델을 선정

#농어 길이
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )

#농어 무게 
perch_weight= np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 
     1000.0, 1000.0]
     )
perch_length.shape, perch_weight.shape
```

```python
### 산점도로 특성들간의 관계 확인하기
import matplotlib.pyplot as plt
```

```python
plt.scatter(perch_length, perch_weight, alpha=0.8) # alpha 투명도 최대=1
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

### 해석
# 독립변수와 종속변수를 x, y 축으로 하는 산점도를 그렸을 때
# 우상향 분포를 나타내고 있습니다.
# 즉, 독립변수 x축의 값이 커질 수록 종속변수 y값도 함께 커지는 관계를 나타내고 있습니다.
# 이러한 관계를 선형관계라고 합니다.
# 예측하고자하는 종속변수가 연속형 데이터인 경우에는 산점도를 이용하여 선형관계가 나타나는지 확인해야합니다.
# 그래프를 볼 때, 초반에는 곡선을 띄는 듯하다가 중반부부터는 직선의 형태를 보이고 있습니다.
# 이처럼 독립변수와 종속변수의 데이터가 선형관계를 나타내는 경우에 사용되는 방법으로는 회귀분석 방법이 있습니다.
```

```python
### 함수 : 리턴이 있는 경우에는 보통 이름앞에 get을 붙이고 첫글자 대문자로 시작
#       : 리턴이 없고 값을 변수에 저장하는 경우의 함수는 set으로 시작
#       : 최초에 변수 값들을 초기화하는 함수의 경우에는 init으로 시작

def getReturn() : 
    a = 10
    b = 12
    c = 33
    return a, b, c
```

```python
aa, bb, cc = getReturn() 
aa, bb, cc
```

```python
### 훈련 및 테스트 데이터로 분류하기
# 훈련:테스트 = 75:25로 분류
# 분류 함수(라이브러리) 사용
# 랜덤 규칙 : 42번 사용
# 변수명 : 훈련(train_input, train_target), 테스트(test_input, test_target)
# 차원 확인
```

```python
from sklearn.model_selection import train_test_split
```

```python
### 훈련:테스트 = 75:25 데이터 분류하기(랜덤)
# stratify : 분류모델에서 종속변수의 범주형 데이터에 대한 편향을 줄이기 위한 속성으로 회귀모델에서는 연속형 데이터를 사용하기에 사용하지 않음
# train_test_split(독립변수, 종속변수, 분할비율, 랜덤규칙)
#   -> 반환되는 값 4개 순서 : 훈련독립, 테스트독립, 훈련종속, 테스트종속
#                          : train_input, test_input, train_target, test_target

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, test_size = 0.25, random_state = 42)
```

```python
(train_input.shape, train_target.shape), (test_input.shape, test_target.shape)
```

```python
### 훈련 및 테스트의 독립변수를 2차원으로 변환하기
# 변환하는 이유 : 머신러닝에서 훈련시에는 2차원 독립변수를 사용합니다.
# 변수명 : train_input_2d, test_input_2d
# reshape(행, 열) : 차원 변경 함수
# reshape(-1, 1) : -1은 모든 행을 의미, 두번째 1은 1열을 의미
#                : 즉, 전체행에 대한 1열을 만드는 작업을 수행합니다.

train_input_2d = train_input.reshape(-1, 1)

test_input_2d = test_input.reshape(-1, 1)

train_input_2d.shape, test_input_2d.shape
```

```python
### 회귀 모델 훈련하기
```

```python
# 회귀 모델(객체) 생성
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
knr
```

```python
# 모델 훈련시키기
# 회귀분석은 지도학습 : 문제(독립변수)와 정답(종속변수)을 넣어 줍니다.
#                    : 문제와 정답은 훈련데이터만 사용합니다.(테스트 데이터는 훈련에 사용하면 안됩니다.)
# fit() : 훈련 또는 학습을 시킨다고 칭합니다.
# 스스로 문제에 대한 정답을 맞추기 위한 패턴을 찾아 냅니다.
# 패턴을 찾은 결과를 모델이 훈련(학습)되었다고 칭합니다.
knr.fit(train_input_2d, train_target)
```

```python
### 훈련시킨 모델로 훈련 및 테스트 정확도 확인하기
# 훈련(학습)된 모델에다가 문제만 넣어서 예측한 후 정답과 비교하는 절차를 거칩니다.
# 에측한 값과 정답과 비교하여 어느정도 잘 맞추었는지를 확률로 나타냅니다.
# score()함수 내부 수행 절차 : 독립변수(문제)를 이용해서 수행하여 예측함
#                          : 예측 결과와 종속변수(정답)와 비교 후 결과를 알려줍니다.

# 훈련 정확도
train_score = knr.score(train_input_2d, train_target)

# 테스트 정확도
test_score = knr.score(test_input_2d, test_target)

train_score, test_score
```

```python
"""
해석
- 훈련 정확도 보다 테스트 정확도가 높다
- 농어 데이터를 이용하여 독립변수 길이와 종속변수 무게에 대한 특성을 이용하여 KNN 회귀분석을 수행한 결과,
  훈련 정확도 및 테스트 정확도는 높게 나왔으나,
  훈련 정확도가 테스트 정확도보다 낮은 값을 보이는 "과소적합"이 발생하였다.
  따라서, 본 KNN 회귀모델을 사용하여 예측하는데는 어려움이 있기에,
  성능 향상을 위한 하이퍼파라메터 튜닝 또는 데이터의 증가가 필요할 것으로 여겨진다.
"""
```

```python
### 모델 평가하기
# 평균 절대 오차(MAE : Mean Absolute Error)
#  : 회귀에서 사용하는 평가 방법 중 주로 사용됨
#  : 예측값과 실제값의 거리 차이를 평균하여 계산한 값을 절대값으로 나타낸 값
#  : 차이가 작을 수록 좋다고 볼 수 있음
```

```python
### 평균절대오차(MAE) 라이브러리 불러들이기
from sklearn.metrics import mean_absolute_error
```

```python
### 테스트 데이터로 예측하기
# MAE에 사용할 예측값은 테스트 데이터로 예측된 값을 의미합니다.
test_pred = knr.predict(test_input_2d)
print(f"test_pred(예측값) = {test_pred}")
print(f"test_target(실제값) = {test_target}")
```

```python
### 평균절대오차(MAE) 확인하기
mea = mean_absolute_error(test_target, test_pred)
mea
```

```python
"""
해석
- KNN 회귀모델을 활용하여 독립변수 길이를 이용하여 무게를 예측하고자 할때
    약 19.157g 정도의 오차가 발생할 수 있는 모델이다.(+- 19.157g)
"""
```

```python
"""
<현재 시점까지의 모델은 과소적합이 발생한 모델>
- 해소하기 위하여 하이퍼파라메터 튜닝을 진행하고자함.
"""
```

```python
### KNN 모델 하이퍼파라메터 튜닝 : 기본적으로 이웃의 갯수(K)값 찾기(가장 적절한 값)
# 하이퍼파라메터의 정의 : 사람이 직접 값을 정해주어야 하는 속성(변수)들을 하이퍼파라메터라고 칭합니다.
# 보통 모델(객체) 생성 시에 하이퍼파라메터 변수들을 지정해 줍니다.
# 최초 모델 훈련 시에는  : 기본값으로 훈련(학습)시킨 후 정확도를 확인합니다.
#                      : 일반적으로 모델이 가지고 있는 기본값(하이퍼파라메터)은 주로 사용되는 기준값이라고 보면 됩니다. 
# 그 다음 정확도를 확인 후 > 과대/과소 적합 여부 확인을 거친 후 > 튜닝이 필요할지 여부를 판단합니다.
```

```python
# Class KNeighborsRegressor :
# def__init__(n=neighbors=5, ..., ...) :
```

```python
# knr = KNeighborsRegressor(n=neighbors=10)
```

```python
### 이웃의 갯수(K) 5개를 직접 정의해서 훈련 및 정확도 확인해보기

### 이웃의 갯수(K) 정의하기
# knr 클래스 내에 n_neighbors 멤버 변수에 접근하여 값을 5로 수정합니다.
knr.n_neighbors = 5

### 훈련(학습) 시키기
knr.fit(train_input_2d, train_target)

### 훈련 정확도 확인하기 : 훈련데이터 사용
train_score = knr.score(train_input_2d, train_target)

### 테스트 정확도 확인하기 : 테스트데이터 사용
test_score = knr.score(test_input_2d, test_target)

train_score, test_score
```

```python
### 이웃의 갯수(K) 3개를 직접 정의해서 훈련 및 정확도 확인해보기

### 이웃의 갯수(K) 정의하기
# knr 클래스 내에 n_neighbors 멤버 변수에 접근하여 값을 5로 수정합니다.
knr.n_neighbors = 3

### 훈련(학습) 시키기
knr.fit(train_input_2d, train_target)

### 훈련 정확도 확인하기 : 훈련데이터 사용
train_score = knr.score(train_input_2d, train_target)

### 테스트 정확도 확인하기 : 테스트데이터 사용
test_score = knr.score(test_input_2d, test_target)

train_score, test_score
```

```python
"""
해석
- 하이퍼파라메터 튜닝을 통해 테스트 정확도는 다소 낮아졌으나,
  훈련 정확도는 높아지는 매우 성능좋은 모델이 되었다.
- 과소 및 과대 적합이 발생하지 않았으며, 훈련 정확도 0.98의 매우 훌륭한 모델로, 일반화된 모델로 사용가능함
"""
```

```python
### 가장 일반화된 성능을 찾기 위한 하이퍼파라메터 찾기 : 자동화

# 정확도가 가장 높을 때의 이웃의 갯수를 담을 변수
nCnt = 0

# 정확도가 가장 높을 때의 정확도를 담을 변수
nScore = 0

for n in range(3, len(train_input_2d), 2):
    # 이웃의 갯수 정의하기
    knr.n_neighbors = n
    
    # 이미 모델 객체가 생성되어 있는 경우에는 score()만 가지고도 정확도 확인 가능
    # fit은 생략가능 -> score()함수 내에서도 fit을 자동 수행하고 있음
    knr.fit(train_input_2d, train_target)
    score = knr.score(train_input_2d, train_target)
    
    # print(n, score)
    # 1보다 작은 정확도인 경우만 처리
    if score < 1:
        # 담아놓은 nScore값이 현재 Score보다 작은 경우에는 현재 Score로 교체하기
        if nScore < score:
            nScore = score
            # 이때의 n(이웃의 갯수)를 nCnt 변수에 담아 놓기
            nCnt = n
            
print(f"nCnt={nCnt}, nScore={nScore}")
```

```python
### 이웃의 갯수 3에 대한 훈련을 한번 수행해야 합니다.
# 위의 하이퍼파라메터 알고리름 수행의 최종 결과값은 반복문의 마지막 결과값만 포함되어 있습니다.
knr.n_neighbors = 3
knr.fit(train_input_2d, train_target)
```

```python
### 이웃의 갯수 3을 이용하여 테스트 데이터 예측하기
test_pred = knr.predict(test_input_2d)
print(f"test_pred(예측값) = {test_pred}")
print(f"test_target(실제값) = {test_target}")
```

```python
### 평가하기 : 평균절대오차(MAE) 확인하기
mea = mean_absolute_error(test_target, test_pred)
mea

### 해석
# 약 35.42g 정도의 오차가 발생할 수 있는 모델이다.(+- 35.42g)
```

```python
"""
<KNN 회귀모델의 한계>
- KNN 모델은 가장 가까운 이웃들의 비율로 종속변수를 예측하기 때문에
- 테스트 데이터와의 거리가 아무리 멀더라도 가장 가까운 값들을 동일하게 찾게되는 현상이 발생함

** 큰 길이를 예측하고자 할 경우에는 해당 모델에 큰길이를 포한하는 데이터를 확보해야 가능하며,
    그렇지 못한 경우에는 해당 KNN g모델은 부적합한 모델로 다른 모델을 찾아봐야 합니다.
"""
```

```python
### 다른 회귀모델 : 선형방정식을 사용함
# 선형회귀모델(Linear Regression Model), 릿지, 라쏘, 앙상블
# 선형회귀모델 : 단일회귀모델(독립변수가 하나인 것, 직선의 방정식), 
#               다항회귀모델(독립변수가 2개인 것, 곡선의 방정식), 
#               다중회귀모델(독립변수가 3개 이상인 것, 단일회귀의 확장)
# 앙상블모델 : 랜덤포레스트, 엑스트라트리, 그레디언부스트, 히스토그램 그레디언트부스트
```

```python
### 단일 선형회귀모델
# 모델 이름 : LinearRegression
```

```python
### 라이브러리 불러블이기
from sklearn.linear_model import LinearRegression
```

```python
### 데이터 확인하기
train_input_2d, train_target, test_input_2d, test_target
train_input_2d.shape, train_target.shape, test_input_2d.shape, test_target.shape
```

```python
### 모델 생성하기
# 모델 변수명 : lr
from sklearn.neighbors import KNeighborsRegressor
lr = LinearRegression()
lr
```

```python
### 모델 훈련 시키기
lr.fit(train_input_2d, train_target)
```

```python
### 훈련 및 테스트 정확도 확인하기
train_score = lr.score(train_input_2d, train_target)
test_score = lr.score(test_input_2d, test_target)

train_score, test_score, train_score - test_score 

### 해석
# 훈련정확도가 0.94로 모델의 성능은 좋으나, 테스트 정확도가 0.825로 훈련정확도와 0.12 정도의 차이를 보이고 있음
# 따라서 과대적합을 의심할 수 있음
```

```python
### 임의 데이터 50cm로 에측하기
pred = lr.predict([[50]])
pred
```

```python
### 원본데이터와 임의데이터 산전도로 시각화하기
plt.scatter(train_input_2d, train_target)
plt.scatter(50, pred[0], marker='^')

plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
### 추세선 그리기
# 추이선, 예측선, 추세선 모두 같은 의미의 용어로 사용됩니다.
# 우리가 분석할 때 추이 분석이라는 용어를 사용할 때 보통 추세선을 의미한다고 보면 됩니다.
# 예측 결과는 추세선 영역 안에서 모델이 예측을 수행합니다.
# 모델이 훈련(fit)을 하게 되면 추세선, 즉 추이(패턴)을 찾기 위한 기준값을 추출합니다.
# 기준값은 : 기울기와 y절편
# 단일 회귀모델 y = ax + b 의 직선의 방정식을 사용함(a는 기울기, b는 y절편)
# 기울기와 절편을 이용하여 예측을 수행하며, 추세선이 그려집니다.
```

```python
### 추세선을 그리기 위해서 계수값(기울기와 절편) 추출하기
# 기울기 : coef_
# y절편 : intercept_

a = lr.coef_
b = lr.intercept_
a, b
```

```python
### 추세선 그리기
### 원본데이터와 임의데이터 산전도로 시각화하기
# 원본
plt.scatter(train_input_2d, train_target)
# 임의 데이터
plt.scatter(50, pred[0], marker='^')
# 추세선(기울기, 절편) : 선 그래프
plt.plot([15, 50], [a*15+b, a*50+b])

plt.xlabel("length")
plt.ylabel("weight")
plt.show()

print(15, a*15+b, 50, a*50+b)
```

```python
"""
해석
- KNN 모델 보다는 좋은 성능을 보이는 모델로,
  과대적합이 의심은 되나, 일반화 되었다고 판단해도 될 것으로 보임
  그러나, y절편의 값이 음수(-)의 값으로, 예측시 0보다 작은 무게를 예측할 수 있는 모델로 여겨짐
- 즉, 0 이하의 "이상치"데이터를 예측할 수 있음
- 따라서, 단일 회귀모델로는 생선의 길이로 무게를 예측하기에는 적합하지 않은 것으로 판단됨

해결 방법
- 원본의 산점도 그래프를 볼 때 시작 부분이 곡선을 보이고 있음
- 이에, 직전이 아닌 곡선이 적용된 모델로 훈련할 필요성이 있음
"""
```

```python
### 다항회귀모델 사용하기
# 곡선의 방정식을 사용함
# y = ax^2 + bx + c
# 독립변수가 2개 사용됨(변수 = 특성)
# 사용모델은 : LinearRegression()
#           : 모든 선형회귀모델(단일, 다항, 다중 모두 동일)은 같은 모델을 사용함
#           : 회귀모델 구분은 독립변수의 형태에 따라서 모델이 알아서 결정함(사람이 개입 안함)
```

```python
train_input_2d.shape
```

```python
### 특성 x^2값을 생성하여 2개의 특성 생성하기
# 사용하는 데이터 : train_input_2d, test_input_2d
import numpy as np

### 훈련 독립변수에 x^2값 생성하여 2열 생성하기
# column_stack() : 행렬의 생성 함수
train_poly = np.column_stack((train_input_2d**2, train_input_2d))
train_poly

### 테스트 독립변수에 x^2값 생성하여 2열 생성하기
test_poly = np.column_stack((test_input_2d**2, test_input_2d))
test_poly
```

```python
### 모델 생성하기
lr = LinearRegression()
lr
```

```python
### 모델 훈련 시키기
lr.fit(train_poly, train_target)
```

```python
### 훈련 및 테스트 정확도 확인하기
train_score = lr.score(train_poly, train_target)
test_score = lr.score(test_poly, test_target)

train_score, test_score, train_score - test_score

### 해석
# 훈련과 테스트의 정확도를 비교 했을 때 과소적합이 미세하게 나타나고 있음
# 매우 미세한 차이의 과소저합인 경우에는 그대로 사용해도 무방하지만, 그래도 성능 향상을 고려해볼 필요성은 있습니다.
# 해소방법 : 데이터 늘이기, 그대로 사용하기(미세한 경우), 다른 모델 찾아서
```

```python
### 임의 값 길이 50cm로 예측하기
# 임의 데이터로 예측 할 때에도 훈련에 사용된 독립변수의 형태와 같게 해야합니다.
pred = lr.predict([[50**2, 50]])
pred
```

```python
### 계수(a, b 기울기 및 c절편) 추출하기
# 기울기
a = lr.coef_[0]
b = lr.coef_[1]

# y절편
c = lr.intercept_

a, b, c
```

```python
### 추세선 그리기
# 원본 데이터
plt.scatter(train_input_2d, train_target)
# 임의 데이터
plt.scatter(50, pred[0], marker="^")
# 추세선 그리기
# 곡선의 x축값 범위 생성하기 : 0~50 범위의 리스트
point = np.arange(0, 51)
plt.plot(point, a*point**2 + b*point + c)

plt.grid()
plt.show()
```

```python
"""
<다항회귀모델>
- 단일회귀모델에 비하여 음의 값이 나올수(예측) 있는 부분을 해소하였으며, 
  예측 결과 또한 정확한 것으로 판단되는 모델임
- 다만, 과소적합이 미세하게 나타나기에 모델로 사용하기에는 부적한 부분이 있음
- 이를 해소하기 위해서는 데이터양 증가 또는 다른 모델을 사용하는 방법을 취할 필요성이 있음

(해소 가능한 방법)
- 과대/과소 적합을 해소할 수 있는 방법으로는 row 데이터를 증가하는 것은 현재 어렵다고 판단되며,
  col(열, 특성) 데이터를 늘리는 방법으로 해볼 수 있음
   -> 특성을 늘린다는 개념은 빅데이터 공학에서 "특성공학" 이라는 개념을 적용함
   -> 특성을 늘림으로써 모델이 더 복잡하게, 즉 집중력을 가지고 패턴을 찾을 수 있도록 해야함
   -> 특성을 늘려서 분석하는 모델로는 다항회귀 모델이 있습니다.(단일 회귀의 확장이라고 생각하면 됨)

- 또 다른 방법
   -> 특성을 늘린다는 의미는 집중력, 즉 복잡도를 늘린다는 개념으로 적용됩니다.
   -> 특성이 많을수록 모델이 패턴을 찾기위해 많은 반복 훈련을 수행 합니다. -> 이를 복잡도라고 생각하면 됨
   - 복잡도라는 속성(하이퍼파라메터)을 이용하여 복잡도를 크게 또는 작게 할 수 있습니다.
   - 복잡도를 적용한 모델이 "릿지"와 "라쏘"가 있습니다.
"""
```
