```python
"""
<앙상블 모델>
- tree 구조(결정트리)를 기반으로 만들어진 모델
- 여러개의 트리 모델을 이용해서 훈련하는 모델을 앙상블 모델이라고 칭합니다.

<앙상블 모델 분류>
- 회귀와 분류에서 모두 사용가능한 앙상블 모델들이 존재합니다.
- 패키지 : sklearn, xgb, 이외 다수
- 배깅(Bagging) 방식의 앙살블 모델과 부스팅(Boosting) 방식의 앙상블 모델로 구분됩니다.
  (내부에서 트리모델 여러개를 어떻게 사용하느지에 따른 구분일 뿐)
"""
```

```python
"""
<배깅(Bagging)과 부스팅(Boosting) 방식>
- 훈련 데이터를 여러 트리모델 각각에 적용하는 방식에 따라서 구분됩니다.

(배깅(Bagging) 방식)
- 사용되는 여러개의 트리모델들 각각은 서로 독립적으로 사용됨(연관성 없음)
- 훈련 데이터는 여러 트리모델 각각에 랜덤하게 적용되며,
    훈련에 사용되는 데이터가 중복되어 적용될 수도 있음
- 훈련데이터를 사용 후에는 반환하는 방식을 사용
    -> 반환 후에 다시 랜덤하게 추출하여 여러 트리모델 각각에 적용됨
       (그래서 훈련 데이터로 중복 사용 될 수 있음)
- 대표적인 배깅 앙상블 모델 : 랜덤포레스트
- 이외 모델 : 엑스트라트리
- 예측 또는 분류 방식 : 회귀에서는 평균값을 이용하며, 분류에서는 과반수(확률)에 따라 결정됨
- *과적합을 해소하는 모델로 주로 사용됨

(부스팅(Boosting) 방식)
- 사용되는 여러개의 트리모델은 서로 연관성을 가집
- 이전 트리모델의 오차를 보완하기 위하여 다음 트리모델에서 이어서 훈련이 진행되는 방식
- 훈련데이터는 랜덤하게 추출되어 사용되지만,
    이전 트리모델에서 오차가 발생했던 훈련데이터에 가중치를 스스로 부여하여 사용하는 방식
- 대표적인 부스팅 앙상블 모델 : 그레디언트부스팅
- 이외 모델 : 히스트그레디언트부스팅, 엑스지부스팅
- 회귀에서는 결정계수, 분류에서는 정확도를 높이기 위해 주로 사용됨
    (과적합 해소에 주로 사용됨)

<실제 모델 적용 순서(규칙은 없음)>
- 앙상블 모델을 이용하는 경우에는 보통 랜덤포레스틀 처음에 진행합니다.(주로 처음에 진행)
    -> 랜덤포레스트 모델은 정확도가 높은 모델로 유명하며, 현재는 부스트 모델이 정확도가 좀 더 높게 나오는 추세임
- *최종 모델 선정시에는 분류모델 전체를 사용하여 비교 후 선정해야 합니다.
"""
```

```python
"""
<데이터 스케일링>
- 서로 다른 변수(특성)의 값의 크기(단위)를 일정한 수준으로 맞추는 작업을 의미함
- 수치형 변수(특성)에만 적용함

<스케일링 방법 3가지>
- 정규화 : StandardScaler, RobustScaler
- 표준화 : MinMaxScaler

* StandardScaler : 분류 또는 회귀에서 사용가능하며, 주로 분류에서 사용됨
* RobustScaler : 분류 및 회귀 모두 사용 가능
* MinMaxScaler : 분류 또는 회귀에서 사용가능하며, 주로 회귀에서 사용됨
                 (최대 및 최소 범위 내에서 처리되는 방식)
* 스케일링의 경우 StandardScaler또는 MinMaxScaler가 주로 사용됨
  (회귀 / 분류 구분하지 않고 모두 적용 후 훈련 성능 비교하면 좋음)
"""
```

```python
### 스케일링 처리를 함수로 만들어서 사용해보기 ---------------------------------------
# 함수이름 : getScaler
# 딕셔너리에 담아서 사용할 겁니다. 딕셔너리 이름 : dict_Scaler
# 딕셔너리 내에 key는 스케일링 이름, value는 스케일링 처리된 train 및 test 데이터를 튜플 형태로
# <내 코드>
def getScaler(train_data, test_data, scaler_type):
    if scaler_type == 'ss':
        scaler = StandardScaler()
    elif scaler_type == 'mm':
        scaler = MinMaxScaler()
    elif scaler_type == 'rs':
        scaler = RobustScaler()
        
    train_scaler = scaler.fit_transform(train_input)
    train_scaler = scaler.transform(test_input)
    
    return train_scaler, train_scaler

dict_Scaler = {
    'ss': getScaler,
    'mm': getScaler,
    'rs': getScaler
}

for scaler_name, scaler_func in dict_Scaler.items():
    scaled_train, scaled_test = scaler_func(train_input, test_input, scaler_name)
    print(f"{scaler_name} scaler:")
    print("Scaled train data:\n", scaled_train)
    print()
    print("Scaled test data:\n", scaled_test)
    print("------------------------------------------------------------")
```

```python
### 스케일링 처리를 함수로 만들어서 사용해보기 ---------------------------------------
# 함수이름 : getScaler
# 딕셔너리에 담아서 사용할 겁니다. 딕셔너리 이름 : dict_Scaler
# 딕셔너리 내에 key는 스케일링 이름, value는 스케일링 처리된 train 및 test 데이터를 튜플 형태로
# <강사님 코드>

def getScaler(scaler, x_train, x_test):
    scaler.fit(train_input)

    x_train_scaler = scaler.transform(x_train)
    x_test_scaler = scaler.transform(x_test)
    
    return x_train_scaler, x_test_scaler
```

```python
ss = StandardScaler()
rs = RobustScaler()
mm = MinMaxScaler()

list_scalers = [ss, rs, mm]

dict_Scaler = {}
for s in list_scalers:
    sName = s.__class__.__name__
    # print(sName)
    # print(getScaler(s, train_input, test_input))
    dict_Scaler[sName] = getScaler(s, train_input, test_input)
    
dict_Scaler['StandardScaler'][0], dict_Scaler['StandardScaler'][1]
```

```python
"""
<분류 모델 평가 방법>
- 정확도, 정밀도, 재현율, F1-Score 값을 이용하여 평가합니다.
- 예측 시 부정확한 오류가 있는지 확인 : 오류는 잘 못 맞춘 경우를 의미합니다.
  (오류는 "오차행렬도(혼동행렬도)"를 통해 확인합니다.)

<평가 기준>
- 정확도를 이용하여 과적합 여부 확인(이미 위에서 진행하였음)
- 평가기준 : 재현율이 높고, F1-Score 값이 높은 모델을 선정하게 됩니다.
          : 재현율이 낮은 경우(0에 가까운 경우)에는 부정확한 오류가 있다는 의미 
- 정확도, 정밀도, 재현율, F1-Score 모든 값의 범위는 0 ~ 1 사이의 값을 가지며, 확률을 의미합니다.
- 값이 높을 수록 좋습니다.(다만, 오차행렬도를 통해 오루 확인을 꼭 해야합니다.)
"""
```

```python
"""
<오차행렬(혼동행렬)>
- 어떠한 유형의 오류가 발생하고 있는지를 나타내는 값
- 이를 시각화한 것을 오차행렬도 또는 혼동행렬도라고 칭함
- 정확도(score)의 값과 오차행렬도의 시각화 결과로 모델을 선정할 수 있음

<해석 방법>
- 긍정(Positive)적 오류인지, 부정(Negative)적 오류인지로 해석됩니다.
- 오류의 종류 : 긍정(Positive)적오류와 부정(Negative)오류로 구분 합니다.

FP(False Positive) : 예측 결과가 맞지는 않음
                     긍정적 오류로 해석함
                     위험하지 않은 오류
                     0을 1으로 예측한 경우
                     
FN(False Negative) : 예측 결과가 맞지는 않음
                     부정적 오류로 해석함
                     위험한 오류
                     1을 0으로 예측한 경우
                     FN의 값이 크다면 -> 정확도(score)의 값이 크더라도 FN값이 큰 경우에는 재현율 값이 낮아지기에,
                     예측모델로 사용하는데 고려해야함(보통 제외시킴)
                     
TP(True Positive) : 예측결과가 맞는 경우
                    1을 1로 예측한 경우

TN(True Negative) : 예측결과가 맞는 경우
                    0을 0으로 예측한 경우
"""
```

```python
"""
<평가에 사용되는 값>
- 정확도, 정밀도, 재현율, F1-Score

*정확도(Accuracy) : 분류모델에서 을 통해 나온 값과 동일합니다.
                    예측결과가 실제값과 얼마나 정확한가를 나타내는 값
                    Accuracy = (TP + TN) / (TP + TN + FP + FN)
                    
*정밀도(Precision) : 모델이 1로 예측한 데이터 중에 실제 1로 잘 예측한 값
                     Precision = TP / (TP + FP)
            
*재현율(Recall) : 실제로 1인 데이터를 1로 잘 예측한 값
                  Recall = TP / (TP + FN)
                  위험한 오류가 포함되어 있음(1을 0으로 판단한 경우)
                  
*F1-Score : 정밀도와 재현율을 조합하여 하나의 통계치로 반환한 값
            F1-Score = (정밀도 * 재현율) /  (정밀도 + 재현율)

<최종 모델 선정 방법>
- 과소 및 과대 적합이 일어나지 않아야함
    : score()를 통해 확인함 -> Accuracy는 별도로 계산하지 않아도 됨
- 재현율과 F1-Score가 모두 높으면 우수함 모델로 평가할 수 있음
- 재현율이 현저히 낮은 경우에는 모델 선정에서 고려할 필요성 있음(제외 대상 모델 또는 성능 향상 필요)
"""
```

```python
### 분류 모델 평가 라이브러리
# 모두 함수 : 사용되는 데이터는 테스트(또는 검증) 종속변수, 예측결과

# 정확도
from sklearn.metrics import accuracy_score
# 정밀도
from sklearn.metrics import precision_score
# 재현율
from sklearn.metrics import recall_score
# F1-Score
from sklearn.metrics import f1_score
```
