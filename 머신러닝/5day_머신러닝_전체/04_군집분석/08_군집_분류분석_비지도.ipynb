```python
"""
<군집(Clustering) 분석>
- 비지도 학습 방법
- 독립변수만 존재하고 종속변수는 존재하지 않는 경우를 비지도 학습이라고 합니다.
- 스터디를 위해 제공되는 데이터인 경우에는 종속변수를 제공해서 예측 결과(군집결과)를 확인 시켜주는 경우도 있습니다.
- 일반적으로는 데이터 수집 시 정제되지 않은 뉴스데이터 이미지데이터 등과 같은 형태의 데이터를 군집분류하게 되기 때문에,
  정답은 없다고 생각하면 됩니다.
- 군집분석은 패턴이 유사한 데이터들간에 묶는 작업을 수행합니다.
- 묶음 단위를 군집(Cluster)라고 칭합니다.
- 대표 모델 : KMeans(K평균) 모델

<KMean(K평균) 군집모델>
- 모델 스스로 패턴이 유사한 특성들을 묶으면서 중심 평균을 구하는 방식
- 점진적으로 패턴이 유사한 특성들끼리 묶어나가는 방식으로
  특성들이 묶이면서 중심에 대한 평균이 이동됨
- 가장 대표적은 군집모델

<특징>
- 군집할 갯수를 미리 정의해야함
- 군집할 갯수는 지도학습에서의 종속변수의 범주의 갯수
- 군집의 갯수를 이미 알고 있는 경우도 있으며, 아닌 경우도 잇음
- 군집의 갯수를 이미 알고 있는 경우에는 직접 갯수를 정의하면 되며,
                    알고 있지 않은 경우에는 하이퍼파라메터 튜닝을 통해 적절한 갯수를 선정하면 됨
                    단, 모델 생성시 최초에 군집의 갯수는 정의해야하기 떄문에 임의 갯수를 정해서 넣어주어야함

<작동 방식>
- 군집의 갯수 정하기 : 군집의 갯수를 K라고 합니다.
- 데이터들을 이용해서 무작위로 K개의 클러스터 중심을 정합니다.(최초 중심점은 무작위로 모델이 스스로 정함)
- 각 데이터들이 가장 가까운 클러스터 중심을 찾아서 배치됩니다.
- 중심점에 모이는 데이터들의 평균값을 지속적으로 계산하여 최적의 중심점을 찾아 이동합니다.
- 클러스터의 중심점에 변화가 없을 때까지 스스로 반복 훈련이 진행됩니다.
"""
```

```python
### 사용할 라이브러리
```

```python
### 데이터 처리를 위해 형식적으로 넣어줍니다.
import pandas as pd
import numpy as np

### 정규화
from sklearn.preprocessing import StandardScaler

### 데이터 분류
from sklearn.model_selection import train_test_split

### 하이퍼파라메터 튜닝
from sklearn.model_selection import GridSearchCV

### 오차행렬도
from sklearn.metrics import accuracy_score, confusion_matrix

### 군집모델
from sklearn.cluster import KMeans
```

```python
### 데이터 불러들이기
```

```python
file_path = "./data/08_wine.csv"
data = pd.read_csv(file_path)
data
```

```python
### 데이터 결측, 이상, 중복 확인
# 이미 사용한 데이터로 확인함
```

```python
### 독립변수와 종속변수 분리하기
# 수집된 데이터 내에 독립변수와 종속변수가 함께 존재하기 때문에, 분리를 합니다.
# 독립변수명 : X(iloc 사용), 종속변수명 : y(컬러명 사용)

X = data.iloc[:, :-1]
print(X.shape)
X
```

```python
y = data["class"]
y.shape
```

```python
### 데이터 스케일링하기
# 클래스 변수 : scaler
# 독립변수명 : X_scaled

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

print(X_scaled.shape)
X_scaled
```

```python
### 훈련 : 테스트 = 8 : 2로 분류하기
# 변수명 : X_train, X_test, y_train, y_test 사용

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"X_train = {X_train}\n X_test = {X_test}\n y_train = {y_train}\n y_test = {y_test}\n")
```

```python
### 하이퍼파라메터 튜닝하기
# 종속변수를 알고있다는 가정하에 진행 / 모르면 진행할 수 없음
```

```python
### 튜닝을 위해 -> 하이퍼파라메터 속성 및 값의 범위 정의하기
# 하이퍼파라메터 변수들 정의
param_grid = {
    # 초기 중심점 설정 횟수
    "n_init" : [10, 30, 50],

    # 최대 반복 횟수
    "max_iter" : [10, 30, 50],

    #중심점의 이동이 설정 값보다 작으면, 종료(설정값 지정, 학습정확도)
    "tol" : [0.0001, 0.001, 0.01, 1]
}

### 튜닝 클래스에 정의할 속성 정의
# 데이터 분할 갯수
cv = 5

#클러스터 갯수(군집할 갯수) 정의
n_cluster = 2

# 평가기준 정의
scoring = "accuracy"
```

```python
### 튜닝 클래스에 넣어 줄 모델 생성하기
# 모델 생성시에 클러스터의 갯수를 지정해 주어야 합니다.

kmeans = KMeans(n_clusters=n_cluster, random_state=42)
kmeans
```

```python
### 튜닝 클래스에 생성한 모델 넣어서 최적의 모델 추출하기
```

```python
### 튜닝 클래스 생성하기
grid_search = GridSearchCV(kmeans, param_grid=param_grid, cv=cv, scoring=scoring)
grid_search
```

```python
### 훈련시키기(튜닝 진행)
grid_search.fit(X_train, y_train)
```

```python
### 최적의 하이퍼파라메터가 적용된 모델 추출하기

best_model = grid_search.best_estimator_
best_model
```

```python
### 훈련 및 테스트 데이터로 예측하기
train_pred = best_model.predict(X_train)
test_pred = best_model.predict(X_test)
```

```python
### 훈련 및 테스트 정확도 확인하기
train_acc = accuracy_score(y_train, train_pred)
test_acc = accuracy_score(y_test, test_pred)

train_acc, test_acc, train_acc - test_acc
```

```python
### 오차행렬 게산하기
conf_matrix = confusion_matrix(y_test, test_pred)
conf_matrix
```

```python
### iris(붓꽃) 데이터셋 라이브러리
from sklearn.datasets import load_iris

### 군집분석 평가 함수
# 실루엣 정확도 : 군집분석 평가 함수
# 값의 범위 : -1 ~ 1 사이의 값 사용
# -1에 가까울수록 : 군집 분류가 잘 안된 경우
# 0에 가까울수록 : 이도 저도 아닌 상태(섞여서 군집된 경우가 많음)
# 1에 가까울수록 : 군집이 잘 된 경
from sklearn.metrics import silhouette_score
```

```python
### iris 데이터 읽어들이기
iris = load_iris()
iris
```

```python
"""
<독립변수 특성>
- sepal length : 꽃받침 길이
- sepal width : 꽃받침 너비
- pental length : 꽃잎 길이
- pental width : 꽃잎 너비
* 단위는 cm

<종속변수 특성>
- 종속변수 범주 : 0, 1, 2로 구성되어 있음(붓꽃의 품종을 의미합니다.)
- 0 : 세토사(setosa)
- 1 : 버시컬러(versicolor)
- 2 : 버지니카(virginica)

<분석 주제>
- 4가지 독립변수 특성을 이용하여 붓꽃의 품종을 군집분류하기
"""
```

```python
### load한 데이터 결과는 디셔너리 타입으로 독립변수와 종속변수 데이터 추출하기
# 독립변수명 : X, 종속변수명 : y
### 머신러닝에서 사용하는 변수명들(규칙은 없습니다. 주로 사용되는 이름들입니다.)
"""
- 독립변수명 : X, X_data, data, 또는 데이터 성격에 맞는 이름
- 종속변수명 : y, y_data, data, 또는 데이터 성격에 맞는 이름

- 훈련독립변수명 :  X_train, train_X, train_input
- 훈련종속변수명 : y_target, train_y, train_target
- 검증독립변수명 : X_val, val_X, val_input
- 검증종속변수명 : y_val, val_y, val_target
- 테스트독립변수명 : X_test, test_X, test_input
- 테스트종속변수명 : y_test, test_y, test_target

* 데이터 분류함수 문법(규칙)
- 함수 작성 규칙 : train_test_split(독립변수, 종속변수, 검증(테스트)기준 분류비율(test_size), 랜덤규칙(random_state))
- 함수가 반환한 값의 순서 : 훈련독립변수, 검증(테스트)독립변수, 훈련종속변수, 검증(테스트)종속변수

"""
```

```python
### iris 데이터에서 독립변수와 종속변수 추출하기
# 독립변수 X, 종속변수 y 사용
X = iris["data"]
y = iris["target"]

len(X), len(y)
```

```python
### 훈련 및 테스트(검증) 데이터로 분류하지 않고, 전체 데이터 모두 사용하여 군집하기
```

```python
### 군집모델 생성하기
# 군집갯수 : 3개
# 초기 중심점 설정 횟수 : 10
# 랜덤규칙번호 : 42
# 군집모델 변수명 : kmeans_model

Kmeans_model = KMeans(n_clusters=3, n_init=10, random_state=42)
Kmeans_model
```

```python
### 모델 훈련하기
# 군집분석에서는 훈련과 예측이 동시에 진행됩니다.
# 훈련 및 예측 함수 : fit_predict()
Kmeans_labels = Kmeans_model.fit_predict(X)
print(Kmeans_labels)
print(y)

### 원본의 0값은 -> 1번 군집으로 분류된 것으로 보임
### 원본의 1값은 -> 0번 군집으로 분류된 것으로 보임
### 원본의 2값은 -> 2번 군집으로 분류된 것으로 보임
# 군집번호는 모델 스스로가 순서 상관없이 자동 부여 시킵니다.
```

```python
### 군집 결과 시각화하기
# 주성분 분석을 통한 시각화(PCA)

### 주성분 분석(PCA)
# 훈련에 사용된 특성들 중에 특징을 가장 잘 나타낼 수 있는 특성들을 추출하는 방식
# 특징을 잘 나타내는 특성을 추출하여, 특성들을 축소 시키는 방식으로 사용됨
# 이를 "차원축소"라고 칭합니다.
# 주요 특징을 가지는 특성들만을 추출하여 사용하는 방식으로, 데이터가 매우 많은 경우에 유리함
# 빠른 처리 가능
```

```python
### 주성분 분석(PCA) 라이브러리
from sklearn.decomposition import PCA
```

```python
### 주성분 분석 클래스 생성하기
# n_components : 주성분을 몇개의 차원으로 축소할지 정의

pca = PCA(n_components=2)
pca
```

```python
### 독립변수에서 주성분 찾기
pca.fit(X)
```

```python
### 찾은 주성분의 차원으로 변환하기
# 주성분의 차원축소로 인해 -> 열의 갯수가 줄어든 것을 볼 수 있습니다.(150행 2열로 변환)
X_pca = pca.transform(X)
print(X_pca.shape)
X_pca
```

```python
### 군집분석 결과 시각화하기
plt.title("KMean")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=Kmeans_labels)
plt.show()
```

```python
### 군집분석 결과 평가하기 : 실루엣 정확도
# -1에 가까울수록 : 군집 분류가 잘 안된 경우
# 0에 가까울수록 : 이도 저도 아닌 상태(섞여서 군집된 경우가 많음)
# 1에 가까울수록 : 군집이 잘 된 경
```

```python
### (군집에 대한 평가 - PCA하기 전에 평가)독립변수와 군집결과 데이트를 이용합니다.
s_score = silhouette_score(X, Kmeans_labels)
s_score
```

```python
### 주성분 2개에 대한 분산비율을 통한 품질 선능 확인하기
# 분산비율의 갯수 추출하기
print(pca.explained_variance_ratio_.shape)

# 분산 비율 데이터 추출하기
print(pca.explained_variance_ratio_)

### 주성분 2개 중에 첫번째 주성분의 데이터(첫번째 열)가 특징을 잘 가지고 있는 데이터임
X_pca

### 주성분 품질성능 데이터 계산하기
# 분산 비율 값을 모두 합산 결과가 품질 성능을 나타냅니다.
# 값의 범위 : 0 ~ 1
# 1에 가까울 수록 품질 성능이 좋다고 판단합니다.
#  (독립변수 전체를 사용하지 않더라도, 주성분 데이터 2개의 특성으로 분류 잘 할 수 있다는 의미)

np.sum(pca.explained_variance_ratio_)
```
