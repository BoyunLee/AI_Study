```python
"""
<앤스콤 시각화>
- 앤스콤이란 분이 순치데이터로 확인하는 것보다 시각적으로 확인하는 것이 더 명확하다는 의견에 따라,
- 예시 데이터를 제공해주고 시각적으로 확인해보는 데이터로,
- 시각화에서 가장 많은 예시로 사용되는 데이터입니다.
- 데이터제공 : seaborn 라이브러리 데이터셋이 포함되어 제공됩니다.
- 데이터 제목 : anscombe
"""
```

```python
import seaborn as sns
```

```python
### anscombe 데이터셋 읽어들이기
ans = sns.load_dataset("anscombe")
ans
```

```python
ans.describe()
```

```python
### 각 범주별로 그룹 찾기
ans_group = ans.groupby(["dataset"])
ans_group.head(2)
ans_group.describe()
```

```python
### 각 dtaset 범주별로 시각화해서 데이터의 분포 확인하기
```

```python
### dataset의 각 범주 데이터를 분리하여 추출하기
data1 = ans[ans["dataset"] == "I"]
data2 = ans[ans["dataset"] == "II"]
data3 = ans[ans["dataset"] == "III"]
data4 = ans[ans["dataset"] == "IV"]

data1, data2, data3, data4
```

```python
### 시각화 라이브러리
import matplotlib.pyplot as plt

### 한글처리
plt.rc("font", family="Malgun Gothic")

### 마이너스 기호처리
plt.rcParams["axes.unicode_minus"] = False
```

```python
### data1에 대한 데이터 분포 확인하기(점으로 표시)
plt.plot(data1["x"], data1["y"], "o")
plt.show()
```

```python
### data2에 대한 데이터 분포 확인하기(점으로 표시)
plt.plot(data2["x"], data2["y"], "o")
plt.show()
```

```python
### data3에 대한 데이터 분포 확인하기(점으로 표시)
plt.plot(data3["x"], data3["y"], "o")
plt.show()
```

```python
### data4에 대한 데이터 분포 확인하기(점으로 표시)
plt.plot(data4["x"], data4["y"], "o")
plt.show()
```

```python
### 하나의 그래프에 4개의 각 그래프를 표현하기
```

```python
### 각 그래프를 하나의 그래프로 표현하기 위해서는
# 전체 그래프를 담당하는 객체가 필요함
# 전체 그래프에 각 그래프를 넣을 공간을 할당해야함
# 4개의 그래프를 넣기 위해서 공간을 4개 만들어주어야 합니다.
# 각 공간은 행열로 공간의 위치를 만들어줍니다.
```

```python
### 그래프 상위객체 변수 정의
fig = plt.figure()
```

```python
### 4개의 각 그래프가 들어갈 공간 만들기
# add_subplot(행, 열, 순번)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

### 각 그래프에 제목 넣기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")

### 각각의 그래프에 데이터 넣기
ax1.plot(data1["x"], data1["y"], "o", c="b")
ax2.plot(data2["x"], data2["y"], "o", c="r")
ax3.plot(data3["x"], data3["y"], "o", c="g")
ax4.plot(data4["x"], data4["y"], "o", c="y")

### 전체 제목 넣기
fig.suptitle("Anscombe Data")

### 겹치는 부분에 대해서는 그래프 재정렬
# 그래프를 모두 완성 후 가장 마지막에 한번만 수행
fig.tight_layout()

### 그래프 그리기
fig
```

```python
### 그래프 저장하기
save_file = "./img/anscombe.png"
fig.savefig(save_file)
```
