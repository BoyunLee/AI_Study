```python
### 라이브러리
# 데이터처리
import pandas as pd

# 시각화 라이브러리, 다양한 데이터셋을 제공하기도 합니다.
import seaborn as sns
```

```python
### 팁(Tip) 데이터셋 활용하기
# toral_bill : 총 힘계 요금표
# tip : 팁
# sex : 성별
# smoker : 흠연 여부
# day : 방문 요일
# time : 방문 시간대
# size : 방문 인원

tips = sns.load_dataset("tips")
tips
```

```python
### tip의 값이 5보다 큰값들 조회
c = tips_del["tip"]>5
c
tips_del[c == True]
```

```python
### 방문요일이 Sun 또는 Sat 이 포함되어 있는 데이터들 조회
c = (tips_del["day"] == "Sun") | (tips_del["day"] == "Sat")
tips_del[c]
```

```python
### 논리 연산 중에 포함관계(on 논리 연산과 동일)를 나타내는 함수
tips[tips["day"].isin(["Sun", "Sat"])]
```

```python
### 컬럼명 sex_no 추가 및 값 입력하기
# sex_no 컬럼 추가시 전체 행의 데이터는 0으로 입력
# sex_no 각 행의 값들을  sex의 값이 Female이면 0으로, Male이면 1으로 수정

tips["sex_no"] = 0
tips.info() #정보 보여주는 코드

tips.loc[tips["sex"] == "Male", "sex_no"] = 1
tips     
```

```python
### 데이터프레임 행단위 정렬하기
# by : 정렬 기준 컬럼(1개 이상 가능)
# axis : 0은 행단위 정렬, 1은 컬럼 단위 정렬
#      : 일반적으로 정렬은 0으로 정렬함 
# ascending=True : 오름차순 정렬
# ascending=False : 내림차순 정렬

tips.sort_values(by=["total_bill"], axis = 0, ascending=False)
```

```python
### 데이터 통합 : 행단위 및 열단위 통합 > 사용되는 함수 concat()
# 행과 열을 구분하는 속성 : axis = (0은 행, 1은 열을 의미)
```

```python
### 샘플 데이터 생성하기
df_temp1 = pd.DataFrame([["Dog", 3], ["Cat", 5], ["Tiger", 2]], index=[0, 1, 2], columns=["동물", "나이"])
df_temp2 = pd.DataFrame([["집", 0], ["집", 0], ["동물원", 0]], index=[0, 1, 2], columns=["사는곳", "뿔갯수"])
df_temp1
df_temp2

### 행단위 데이터 통합하기
# axis=0 열로 합치기
# axis=1 행으로 합치기
pd.concat([df_temp1, df_temp2], axis=1)
```
